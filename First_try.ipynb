```python
import os
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader



import torchvision
from torchvision.models import resnet50, ResNet50_Weights
import torchvision.transforms as transforms
from torchvision.ops import nms

import math
import json
import cv2
from PIL import Image
from sklearn.metrics import roc_auc_score, mean_absolute_error, r2_score
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
from typing import Tuple, Dict, List, Optional
import warnings
import re
from glob import glob
import shutil
warnings.filterwarnings('ignore')
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score, f1_score

import csv



# MONAI imports
from monai.networks.nets import UNet
from monai.networks.layers import Norm
from monai.data import Dataset as MonaiDataset, DataLoader as MonaiDataLoader
from monai.transforms import (
    Compose, LoadImaged, EnsureChannelFirstd, Spacingd, Orientationd,
    ScaleIntensityRanged, CropForegroundd, Resized, EnsureTyped,
    Activationsd, AsDiscreted, SaveImaged
)
from sklearn.model_selection import StratifiedShuffleSplit
from monai.losses import DiceLoss, DiceCELoss
from monai.metrics import DiceMetric
from monai.inferers import sliding_window_inference
from monai.data import decollate_batch
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
import hashlib
from collections import Counter
from tqdm.auto import tqdm, trange

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
if torch.cuda.is_available():
    print(f"[GPU] Using: {torch.cuda.get_device_name(0)}")
else:
    print("[GPU] GPU not available; running on CPU.")


import random
def set_seed(seed=42):
    random.seed(seed); np.random.seed(seed)
    torch.manual_seed(seed); torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(42)
torch.backends.cudnn.benchmark = True           # autotune best conv kernels
torch.backends.cudnn.deterministic = False      # allow fastest non-deterministic algorithms
torch.backends.cuda.matmul.allow_tf32 = True    # TF32 for matmul
torch.backends.cudnn.allow_tf32 = True          # TF32 for cuDNN convs
try:
    torch.set_float32_matmul_precision("high")  # PyTorch 2.x
except Exception:
    pass



# =============================================================================
# DATASET LOADING FUNCTIONS
# =============================================================================

class MultiDatasetLoader:
    def __init__(self, dataset_root: str, results_dir: str):
        self.dataset_root = dataset_root
        self.results_dir = results_dir
        self.all_samples = []
        print(f"🔍 Dataset root: {self.dataset_root}")
        print(f"📁 Available folders: {os.listdir(self.dataset_root) if os.path.exists(self.dataset_root) else 'Root not found!'}")



    @staticmethod
    def build_chaksu_labels_from_csvs(csv_paths: List[str]) -> Dict[str, int]:
        """
        Read Bosch/Forus/Remidio 'majority' CSVs and produce a {basename -> label} map.
        Label rules: NORMAL -> 0; any 'GLAUCOMA' (including 'GLAUCOMA SUSPECT') -> 1.
        If the same filename appears in multiple CSVs, take a majority vote across CSVs.
        Ties default to 1 (risk-aware).
        """
        # Collect votes per basename
        votes: Dict[str, List[int]] = {}

        def parse_label(raw: str) -> Optional[int]:
            if raw is None:
                return None
            s = str(raw).strip().upper().replace("SUSUPECT", "SUSPECT")
            if s == "NORMAL":
                return 0
            if "GLAUCOMA" in s:           # includes 'GLAUCOMA SUSPECT'
                return 1
            return None

        # Accept column variants for robustness
        POSSIBLE_LABEL_COLS = {"majority decision", "glaucoma decision", "label", "final label"}
        POSSIBLE_IMAGE_COLS = {"images", "image", "filename", "file", "path"}

        for p in csv_paths:
            if not os.path.exists(p):
                print(f"⚠️ Missing CSV: {p}")
                continue
            try:
                df = pd.read_csv(p)
            except Exception as e:
                print(f"⚠️ Skipping {p}: {e}")
                continue

            # Locate columns
            cols_lower = {c.lower().strip(): c for c in df.columns}
            label_col_name = next((cols_lower[c] for c in POSSIBLE_LABEL_COLS if c in cols_lower), None)
            image_col_name = next((cols_lower[c] for c in POSSIBLE_IMAGE_COLS if c in cols_lower), None)
            if label_col_name is None or image_col_name is None:
                print(f"⚠️ {p}: required columns not found (have: {list(df.columns)}).")
                continue

            for _, row in df.iterrows():
                lab = parse_label(row[label_col_name])
                if lab is None:
                    continue
                images_field = str(row[image_col_name])

                # Extract every filename with a valid image extension (handles 'A.jpg-A-1.jpg' etc.)
                files = re.findall(r'[^,\s]+?\.(?:jpg|jpeg|png|tif|tiff)', images_field, flags=re.IGNORECASE)
                if not files:
                    # fall back: if the cell itself is a single filename
                    if re.search(r'\.(jpg|jpeg|png|tif|tiff)$', images_field, flags=re.IGNORECASE):
                        files = [images_field]

                for f in files:
                    base = os.path.basename(f).lstrip("-_ ").strip()
                    votes.setdefault(base, []).append(lab)

        # Finalize majority per basename
        mapping: Dict[str, int] = {}
        for base, lst in votes.items():
            pos = sum(lst)
            neg = len(lst) - pos
            if pos > neg:
                mapping[base] = 1
            elif neg > pos:
                mapping[base] = 0
            else:
                mapping[base] = 1  # tie → positive (documented above)

        print(f"🗳️ Aggregated votes for {len(mapping)} filenames (from {len(csv_paths)} CSVs).")
        return mapping



    def _first_existing(self, *candidates):
        """Find first existing folder from candidates"""
        for rel in candidates:
            p = os.path.join(self.dataset_root, rel)
            if os.path.exists(p):
                print(f"✅ Found: {rel}")
                return rel
        print(f"❌ None found from: {candidates}")
        return None

    def _find_folder_containing(self, *keywords):
        """Find folder containing any of the keywords (case-insensitive)"""
        if not os.path.exists(self.dataset_root):
            return None

        for folder in os.listdir(self.dataset_root):
            folder_lower = folder.lower()
            for keyword in keywords:
                if keyword.lower() in folder_lower:
                    print(f"✅ Found folder containing '{keyword}': {folder}")
                    return folder
        print(f"❌ No folder found containing: {keywords}")
        return None

    def load_drishti_gs_disc_only(self):
        """Load Drishti-GS dataset with various folder structure possibilities"""
        print("📦 Loading Drishti-GS (disc masks only)...")

        # Try various folder name patterns
        drishti_folder = self._first_existing(
            "Drishti-GS", "DrishtiGS", "Drishti-GS1", "drishti-gs", "DRISHTI-GS"
        )

        if not drishti_folder:
            drishti_folder = self._find_folder_containing("drishti")

        if not drishti_folder:
            print("⚠️ Drishti-GS folder not found")
            return []

        drishti_root = os.path.join(self.dataset_root, drishti_folder)
        print(f"📂 Drishti folder contents: {os.listdir(drishti_root)}")

        # Try different image/GT folder structures
        img_candidates = [
            os.path.join(drishti_root, "Images"),
            os.path.join(drishti_root, "Training"),
            os.path.join(drishti_root, "images"),
            drishti_root  # Sometimes images are directly in root
        ]

        gt_candidates = [
            os.path.join(drishti_root, "GT"),
            os.path.join(drishti_root, "Groundtruths"),
            os.path.join(drishti_root, "gt"),
            os.path.join(drishti_root, "GroundTruth")
        ]

        img_root = None
        for candidate in img_candidates:
            if os.path.exists(candidate):
                img_root = candidate
                print(f"✅ Images found at: {candidate}")
                break

        gt_root = None
        for candidate in gt_candidates:
            if os.path.exists(candidate):
                gt_root = candidate
                print(f"✅ GT found at: {candidate}")
                break

        if not img_root or not gt_root:
            print("⚠️ Could not find Images or GT folders in Drishti-GS")
            return []

        samples = []
        total_images = 0
        loaded_pairs = 0

        # Try both structured (GLAUCOMA/NORMAL folders) and flat structure
        label_folders = ["GLAUCOMA", "NORMAL", "glaucoma", "normal"]
        found_structured = False

        # Check for structured approach first
        for label_folder in label_folders:
            img_folder = os.path.join(img_root, label_folder)
            if os.path.exists(img_folder):
                found_structured = True
                label = 1 if label_folder.upper() == "GLAUCOMA" else 0

                for img_path in glob(os.path.join(img_folder, "*.png")):
                    total_images += 1
                    base_id = os.path.splitext(os.path.basename(img_path))[0]

                    # Try multiple mask patterns
                    mask_patterns = [
                        os.path.join(gt_root, base_id, "SoftMap", f"{base_id}_ODsegSoftmap.png"),
                        os.path.join(gt_root, f"{base_id}_ODsegSoftmap.png"),
                        os.path.join(gt_root, f"{base_id}_disc.png"),
                        os.path.join(gt_root, f"{base_id}.png"),
                    ]

                    mask_found = False
                    for mask_pattern in mask_patterns:
                        if os.path.exists(mask_pattern):
                            samples.append({
                                "image": img_path,
                                "mask": mask_pattern,
                                "label": label,
                                "dataset": "Drishti-GS"
                            })
                            loaded_pairs += 1
                            mask_found = True
                            break

                    if not mask_found:
                        print(f"⚠️ No mask found for {base_id}")

        # If no structured folders, try flat structure with filename-based labeling
        if not found_structured:
            print("🔍 Trying flat structure with filename-based labeling...")
            for img_path in glob(os.path.join(img_root, "**/*.png"), recursive=True):
                total_images += 1
                base_id = os.path.splitext(os.path.basename(img_path))[0]

                # Infer label from path or filename
                path_lower = img_path.lower()
                if "glaucoma" in path_lower or "g_" in base_id.lower():
                    label = 1
                elif "normal" in path_lower or "n_" in base_id.lower():
                    label = 0
                else:
                    continue  # Skip unlabeled

                # Try multiple mask patterns
                mask_patterns = [
                    os.path.join(gt_root, base_id, "SoftMap", f"{base_id}_ODsegSoftmap.png"),
                    os.path.join(gt_root, f"{base_id}_ODsegSoftmap.png"),
                    os.path.join(gt_root, f"{base_id}_disc.png"),
                    os.path.join(gt_root, f"{base_id}.png"),
                ]

                for mask_pattern in mask_patterns:
                    if os.path.exists(mask_pattern):
                        samples.append({
                            "image": img_path,
                            "mask": mask_pattern,
                            "label": label,
                            "dataset": "Drishti-GS"
                        })
                        loaded_pairs += 1
                        break

        print(f"✅ Drishti-GS: Loaded {loaded_pairs} pairs from {total_images} images")
        return samples

    def load_rimone_dl_disc_only(self):
        """Load RIM-ONE DL dataset"""
        print("📦 Loading RIM-ONE DL...")

        # Try various folder name patterns
        rimone_folder = self._first_existing(
            "RIM-ONE-DL", "RIM-ONE", "rim_one", "rimone", "RIM_ONE_DL"
        )

        if not rimone_folder:
            rimone_folder = self._find_folder_containing("rim", "one")

        if not rimone_folder:
            print("⚠️ RIM-ONE folder not found")
            return []

        rimone_root = os.path.join(self.dataset_root, rimone_folder)
        print(f"📂 RIM-ONE folder contents: {os.listdir(rimone_root)}")

        # Try different structures
        img_candidates = [
            os.path.join(rimone_root, "Images"),
            os.path.join(rimone_root, "images"),
            os.path.join(rimone_root, "Image")
        ]

        gt_candidates = [
            os.path.join(rimone_root, "GT"),
            os.path.join(rimone_root, "gt"),
            os.path.join(rimone_root, "Masks"),
            os.path.join(rimone_root, "masks")
        ]

        img_root = None
        for candidate in img_candidates:
            if os.path.exists(candidate):
                img_root = candidate
                print(f"✅ Images found at: {candidate}")
                break

        gt_root = None
        for candidate in gt_candidates:
            if os.path.exists(candidate):
                gt_root = candidate
                print(f"✅ GT found at: {candidate}")
                break

        if not img_root or not gt_root:
            print("⚠️ Could not find Images or GT folders in RIM-ONE")
            return []

        samples = []
        total_images = 0
        loaded_pairs = 0

        # Try structured approach
        for label_folder in ["GLAUCOMA", "NORMAL", "glaucoma", "normal"]:
            img_folder = os.path.join(img_root, label_folder)
            gt_folder = os.path.join(gt_root, label_folder.lower())  # GT often lowercase

            if os.path.exists(img_folder):
                label = 1 if label_folder.upper() == "GLAUCOMA" else 0

                for img_path in glob(os.path.join(img_folder, "*.png")):
                    total_images += 1
                    base_name = os.path.splitext(os.path.basename(img_path))[0]

                    # Try multiple mask patterns
                    mask_patterns = [
                        os.path.join(gt_folder, f"{base_name}-1-Disc-T.png"),
                        os.path.join(gt_folder, f"{base_name}-Disc-T.png"),
                        os.path.join(gt_folder, f"{base_name}_disc.png"),
                        os.path.join(gt_folder, f"{base_name}.png"),
                    ]

                    for mask_pattern in mask_patterns:
                        if os.path.exists(mask_pattern):
                            samples.append({
                                "image": img_path,
                                "mask": mask_pattern,
                                "label": label,
                                "dataset": "RIM-ONE-DL"
                            })
                            loaded_pairs += 1
                            break

        print(f"✅ RIM-ONE DL: Loaded {loaded_pairs} pairs from {total_images} images")
        return samples

    def load_g1020_split(self):
        """Load G1020 split dataset"""
        print("📦 Loading G1020...")

        g1020_folder = self._first_existing("G1020_split", "G1020", "g1020")
        if not g1020_folder:
            g1020_folder = self._find_folder_containing("g1020")

        if not g1020_folder:
            print("⚠️ G1020 folder not found")
            return []

        g1020_root = os.path.join(self.dataset_root, g1020_folder)
        print(f"📂 G1020 folder contents: {os.listdir(g1020_root)}")

        samples = []
        # Try canonical split layout first
        for label_folder in ["glaucoma", "control"]:
            for sub_images in ["images", "Images"]:
                for sub_masks in ["masks", "Masks"]:
                    img_dir = os.path.join(g1020_root, label_folder, sub_images)
                    msk_dir = os.path.join(g1020_root, label_folder, sub_masks)
                    if os.path.isdir(img_dir) and os.path.isdir(msk_dir):
                        print(f"✅ Found G1020 structure: {label_folder}/{sub_images} & {sub_masks}")
                        for img_path in glob(os.path.join(img_dir, "*")):
                            base = os.path.splitext(os.path.basename(img_path))[0]
                            mask_path = os.path.join(msk_dir, f"{base}.png")
                            if os.path.exists(mask_path):
                                samples.append({
                                    "image": img_path,
                                    "mask": mask_path,
                                    "label": 1 if label_folder == "glaucoma" else 0,
                                    "dataset": "G1020"
                                })

        print(f"✅ G1020: Loaded {len(samples)} pairs")
        return samples

    def load_origa_split(self):
        """Load ORIGA split dataset"""
        print("📦 Loading ORIGA...")

        origa_folder = self._first_existing("ORIGA_split", "ORIGA", "origa")
        if not origa_folder:
            origa_folder = self._find_folder_containing("origa")

        if not origa_folder:
            print("⚠️ ORIGA folder not found")
            return []

        origa_root = os.path.join(self.dataset_root, origa_folder)
        print(f"📂 ORIGA folder contents: {os.listdir(origa_root)}")

        samples = []
        for label_folder in ["glaucoma", "control"]:
            for sub_images in ["images", "Images"]:
                for sub_masks in ["masks", "Masks"]:
                    img_dir = os.path.join(origa_root, label_folder, sub_images)
                    msk_dir = os.path.join(origa_root, label_folder, sub_masks)
                    if os.path.isdir(img_dir) and os.path.isdir(msk_dir):
                        print(f"✅ Found ORIGA structure: {label_folder}/{sub_images} & {sub_masks}")
                        for img_path in glob(os.path.join(img_dir, "*")):
                            base = os.path.splitext(os.path.basename(img_path))[0]
                            mask_path = os.path.join(msk_dir, f"{base}.png")
                            if os.path.exists(mask_path):
                                samples.append({
                                    "image": img_path,
                                    "mask": mask_path,
                                    "label": 1 if label_folder == "glaucoma" else 0,
                                    "dataset": "ORIGA"
                                })

        print(f"✅ ORIGA: Loaded {len(samples)} pairs")
        return samples

    def load_refuge(self):
        """Load REFUGE dataset with JPG images and BMP masks"""
        print("📦 Loading REFUGE...")

        refuge_folder = self._first_existing("Refuge", "REFUGE", "refuge")
        if not refuge_folder:
            refuge_folder = self._find_folder_containing("refuge")

        if not refuge_folder:
            print("⚠️ REFUGE folder not found")
            return []

        refuge_root = os.path.join(self.dataset_root, refuge_folder)
        print(f"📂 REFUGE folder contents: {os.listdir(refuge_root)}")

        image_root = os.path.join(refuge_root, "Images")
        gt_root = os.path.join(refuge_root, "GT")

        if not os.path.exists(image_root) or not os.path.exists(gt_root):
            print("⚠️ REFUGE Images or GT folder not found")
            return []

        samples = []
        total_images = 0
        loaded_pairs = 0

        def _find_with_any_ext(folder, base, exts):
            for ext in exts:
                for case_ext in [ext, ext.upper()]:
                    p = os.path.join(folder, base + case_ext)
                    if os.path.exists(p):
                        return p
            return None

        label_folders = ["GLAUCOMA", "NORMAL"]
        label_map = {"GLAUCOMA": 1, "NORMAL": 0}

        for label_folder in label_folders:
            img_dir = os.path.join(image_root, label_folder)
            mask_dir = os.path.join(gt_root, label_folder)

            if not os.path.exists(img_dir) or not os.path.exists(mask_dir):
                continue

            # Find all JPG/JPEG images
            img_patterns = [
                os.path.join(img_dir, "*.jpg"),
                os.path.join(img_dir, "*.jpeg"),
                os.path.join(img_dir, "*.JPG"),
                os.path.join(img_dir, "*.JPEG")
            ]

            img_paths = []
            for pattern in img_patterns:
                img_paths.extend(glob(pattern))

            for img_path in img_paths:
                total_images += 1
                base = os.path.splitext(os.path.basename(img_path))[0]

                mask_path = _find_with_any_ext(mask_dir, base, [".bmp", ".BMP"])

                if mask_path and os.path.exists(mask_path):
                    samples.append({
                        "image": img_path,
                        "mask": mask_path,
                        "label": label_map[label_folder],
                        "dataset": "REFUGE"
                    })
                    loaded_pairs += 1

        print(f"✅ REFUGE: Loaded {loaded_pairs} pairs from {total_images} images")
        return samples


    def _build_chaksu_split_fallback_map(self) -> Dict[str, int]:
        """
        Build a fallback {normalized_basename -> label} map from Chaksu_split/
          expected layout:
            Chaksu_split/
              ├── glaucoma/images/*.png (or jpg/jpeg)
              └── control/images/*.png  (or jpg/jpeg)
        Returns empty dict if the split folder isn't present.
        """
        split_folder = self._first_existing("Chaksu_split", "chaksu_split")
        if not split_folder:
            return {}

        root = os.path.join(self.dataset_root, split_folder)
        mapping: Dict[str, int] = {}

        def _norm(name: str) -> str:
            stem = os.path.splitext(os.path.basename(name))[0].lower()
            return re.sub(r'[^a-z0-9]+', '', stem)

        for label_name, lab in [("glaucoma", 1), ("control", 0)]:
            img_dir = os.path.join(root, label_name, "images")
            if not os.path.isdir(img_dir):
                continue
            for ext in ("*.png", "*.jpg", "*.jpeg", "*.PNG", "*.JPG", "*.JPEG"):
                for p in glob(os.path.join(img_dir, ext)):
                    mapping[_norm(p)] = lab

        print(f"[DBG] Chaksu_split fallback map size: {len(mapping)}")
        # Show a few examples
        shown = 0
        for k, v in mapping.items():
            print(f"[DBG] split_map: {k} -> {v}")
            shown += 1
            if shown >= 5:
                break
        return mapping


    def load_chaksu_img_gt(self):
        """
        Load Chaksu with:
          Chaksu/
            ├─ IMG/   (images: .png/.jpg/.jpeg)
            └─ GT/    (masks; same stem as image; allow .tif/.tiff/.png)
          Labels come from three CSVs inside Chaksu/:
            - Glaucoma_Decision_Comparison_Bosch_majority.csv
            - Glaucoma_Decision_Comparison_Forus_majority.csv
            - Glaucoma_Decision_Comparison_Remidio_majority.csv

        CSV columns (case-insensitive):
          'Images' → filename(s)
          'Majority Decision' → 'NORMAL' → 0, any string containing 'GLAUCOMA' → 1
        Ties across CSVs default to 1 (risk-aware).
        """
        print("📦 Loading Chaksu (IMG/ + GT/ + majority CSVs)...")

        chaksu_folder = self._first_existing("Chaksu", "chaksu")
        if not chaksu_folder:
            chaksu_folder = self._find_folder_containing("chaksu")
        if not chaksu_folder:
            print("⚠️ Chaksu folder not found")
            return []

        root = os.path.join(self.dataset_root, chaksu_folder)
        print(f"📂 Chaksu folder contents: {os.listdir(root)}")

        img_dir = os.path.join(root, "IMG")
        gt_dir  = os.path.join(root, "GT")
        if not (os.path.isdir(img_dir) and os.path.isdir(gt_dir)):
            print("⚠️ Missing IMG/ or GT/ inside Chaksu/")
            return []

        # --- read labels from the three CSVs ---
        csv_names = [
            "Glaucoma_Decision_Comparison_Bosch_majority.csv",
            "Glaucoma_Decision_Comparison_Forus_majority.csv",
            "Glaucoma_Decision_Comparison_Remidio_majority.csv",
        ]
        csv_paths = [os.path.join(root, n) for n in csv_names if os.path.exists(os.path.join(root, n))]
        if not csv_paths:
            print("⚠️ No majority CSVs found in Chaksu/. Proceeding without labels.")
            votes = {}
        else:
            votes: Dict[str, List[int]] = {}

            def parse_lab(s: str) -> int | None:
                if s is None:
                    return None
                t = str(s).strip().upper()
                if t == "NORMAL":
                    return 0
                if "GLAUCOMA" in t:   # includes 'GLAUCOMA SUSPECT'
                    return 1
                return None

            for p in csv_paths:
                try:
                    df = pd.read_csv(p)
                except Exception as e:
                    print(f"⚠️ Skipping CSV {p}: {e}")
                    continue

                cols = {c.lower().strip(): c for c in df.columns}
                img_col = cols.get("images") or cols.get("image") or None
                dec_col = cols.get("majority decision") or cols.get("glaucoma decision") or None
                if img_col is None or dec_col is None:
                    print(f"⚠️ {os.path.basename(p)} missing 'Images'/'Majority Decision' columns.")
                    continue

                for _, r in df.iterrows():
                    lab = parse_lab(r.get(dec_col))
                    if lab is None:
                        continue
                    images_field = str(r.get(img_col, "")).strip()
                    # A cell may contain one or multiple names; accept comma/space/dash separated + extensions
                    files = re.findall(r'[^,\s]+?\.(?:jpg|jpeg|png|tif|tiff)', images_field, flags=re.IGNORECASE)
                    if not files:
                        if re.search(r'\.(jpg|jpeg|png|tif|tiff)$', images_field, flags=re.IGNORECASE):
                            files = [images_field]
                    for f in files:
                        base = os.path.basename(f).lstrip("-_ ").strip()
                        votes.setdefault(base, []).append(int(lab))

            # majority map
            label_map: Dict[str, int] = {}
            for base, lst in votes.items():
                pos = sum(lst); neg = len(lst) - pos
                label_map[base] = 1 if pos >= neg else 0
            print(f"🗳️ Chaksu: aggregated labels for {len(label_map)} filenames from {len(csv_paths)} CSV(s).")

        # helper: find mask for image stem with flexible extensions (prefer .tif/.tiff)
        def _find_mask_for(stem: str) -> Optional[str]:
            exts = [".tif", ".tiff", ".png", ".jpg", ".jpeg"]
            for ext in exts:
                for case_ext in (ext, ext.upper()):
                    p = os.path.join(gt_dir, stem + case_ext)
                    if os.path.exists(p):
                        return p
            return None

        # build a more robust lookup: exact filename and normalized stem
        def _norm(name: str) -> str:
            return re.sub(r'[^a-z0-9]+', '', os.path.splitext(os.path.basename(name))[0].lower())

        label_by_exact = {os.path.basename(k).lower(): v for k, v in (label_map.items() if 'label_map' in locals() else {})}
        label_by_norm  = { _norm(k): v for k, v in (label_map.items() if 'label_map' in locals() else {}) }

        # iterate images in IMG
        img_paths = []
        for pat in ("*.png", "*.jpg", "*.jpeg", "*.PNG", "*.JPG", "*.JPEG"):
            img_paths.extend(glob(os.path.join(img_dir, pat)))

        samples = []
        for ip in img_paths:
            stem = os.path.splitext(os.path.basename(ip))[0]
            mp = _find_mask_for(stem)
            if not mp:
                continue

            # choose label (prefer exact filename match, then norm)
            y = None
            if label_by_exact:
                y = label_by_exact.get(os.path.basename(ip).lower(), None)
            if y is None and label_by_norm:
                y = label_by_norm.get(_norm(ip), None)

            if y is None:
                # if labels are mandatory, skip; else set -1
                # Here we require labels, so skip unlabeled
                continue

            samples.append({
                "image": ip,
                "mask": mp,
                "label": int(y),
                "dataset": "Chaksu"
            })

        print(f"✅ Chaksu: Loaded {len(samples)} image+mask pairs with labels from IMG/ + GT/.")
        return samples


    def _expand_numeric_keys(self, label_map: Dict[str, int]) -> Dict[str, int]:
        """
        If keys are numeric like '1', also add common filename-style variants:
        v0001, 0001, val0001, refuge0001, etc.
        """
        expanded = dict(label_map)
        for k, v in list(label_map.items()):
            if k.isdigit():
                n = int(k)
                variants = {
                    f"{n}", f"{n:03d}", f"{n:04d}",
                    f"v{n}", f"v{n:03d}", f"v{n:04d}",
                    f"val{n}", f"val{n:03d}", f"val{n:04d}",
                    f"refuge{n}", f"refuge{n:03d}", f"refuge{n:04d}",
                }
                for kk in variants:
                    expanded.setdefault(kk, v)
        return expanded



    def _norm_stem(self, name: str) -> str:
        stem = os.path.splitext(os.path.basename(str(name)))[0].lower()
        return re.sub(r'[^a-z0-9]+', '', stem)

    def _parse_binary_label(self, raw) -> Optional[int]:
        if pd.isna(raw):
            return None
        s = str(raw).strip().lower()
        if s in {"1","glaucoma","g","positive","pos","abnormal","glaucoma suspect","suspect"}:
            return 1
        if s in {"0","normal","n","negative","neg","control"}:
            return 0
        try:
            v = int(float(s))
            if v in (0,1):
                return v
        except Exception:
            pass
        return None


    def _file_md5(self, path: str, chunk: int = 1 << 20) -> str:
        import hashlib
        h = hashlib.md5()
        with open(path, "rb") as f:
            for b in iter(lambda: f.read(chunk), b""):
                h.update(b)
        return h.hexdigest()

    def _build_chaksu_split_fallback_maps(self):
        """
        Build multiple fallbacks from Chaksu_split:
          - norm_map:  {normalized_basename -> label}
          - size_map:  {filesize_bytes -> label}         (use only if unique)
          - md5_map:   {md5hex -> label}
        Returns (norm_map, size_map, md5_map)
        """
        split_folder = self._first_existing("Chaksu_split", "chaksu_split")
        if not split_folder:
            return {}, {}, {}

        root = os.path.join(self.dataset_root, split_folder)
        norm_map, size_bucket, md5_map = {}, {}, {}

        def _norm(name: str) -> str:
            stem = os.path.splitext(os.path.basename(name))[0].lower()
            return re.sub(r'[^a-z0-9]+', '', stem)

        total = 0
        for label_name, lab in [("glaucoma", 1), ("control", 0)]:
            img_dir = os.path.join(root, label_name, "images")
            if not os.path.isdir(img_dir):
                continue
            for ext in ("*.png", "*.jpg", "*.jpeg", "*.PNG", "*.JPG", "*.JPEG"):
                for p in glob(os.path.join(img_dir, ext)):
                    total += 1
                    # name-based
                    norm_map[_norm(p)] = lab
                    # size-based (bucket first; we’ll filter non-unique later)
                    try:
                        sz = os.path.getsize(p)
                        size_bucket.setdefault(sz, set()).add(lab)
                    except Exception:
                        pass
                    # md5-based
                    try:
                        md5_map[self._file_md5(p)] = lab
                    except Exception:
                        pass

        # keep only sizes that map to a single label to avoid collisions
        size_map = {sz: next(iter(labels)) for sz, labels in size_bucket.items() if len(labels) == 1}

        print(f"[DBG] Chaksu_split fallback maps: norm={len(norm_map)}, size={len(size_map)} (unique), md5={len(md5_map)}, total_scanned={total}")
        # A few samples:
        for i, (k, v) in enumerate(norm_map.items()):
            print(f"[DBG] split_norm: {k} -> {v}")
            if i >= 4: break
        for i, (k, v) in enumerate(size_map.items()):
            print(f"[DBG] split_size: {k} -> {v}")
            if i >= 4: break
        for i, (k, v) in enumerate(md5_map.items()):
            print(f"[DBG] split_md5: {k[:10]}... -> {v}")
            if i >= 4: break

        return norm_map, size_map, md5_map



    def _load_label_map_from_excel(self, xlsx_path: str,
                                  image_col_candidates=None,
                                  label_col_name="Glaucoma Label",
                                  known_img_basenames: Optional[set] = None) -> Dict[str, int]:
        if image_col_candidates is None:
            image_col_candidates = {
                "image","images","img","filename","file","path","name","id",
                "image name","image_name","img_name","imageid","image id",
                "imagename","img id","imgid"
            }

        if not os.path.exists(xlsx_path):
            raise FileNotFoundError(f"Excel not found: {xlsx_path}")

        xl = pd.ExcelFile(xlsx_path)
        label_map = {}
        found = False
        label_col_lower = label_col_name.strip().lower()

        known_norm = set(known_img_basenames or ())

        for sheet in xl.sheet_names:
            df = xl.parse(sheet)
            cols_lower = {c.lower().strip(): c for c in df.columns}
            if label_col_lower not in cols_lower:
                continue

            cand_cols = [cols_lower[c] for c in image_col_candidates if c in cols_lower]

            chosen_image_col = None
            best_overlap = -1

            def _col_overlap(col_name: str) -> int:
                series = df[col_name].dropna().astype(str)
                series = series[series.str.len() > 0]
                norms = { self._norm_stem(v) for v in series.values.tolist() }
                return len(norms & known_norm), norms

            norms_for_chosen = set()
            if cand_cols:
                for c in cand_cols:
                    try:
                        overlap, norms = _col_overlap(c)
                    except Exception:
                        continue
                    if overlap > best_overlap:
                        best_overlap = overlap
                        chosen_image_col = c
                        norms_for_chosen = norms

            if chosen_image_col is None:
                for c in df.columns:
                    if c == cols_lower[label_col_lower]:
                        continue
                    try:
                        overlap, norms = _col_overlap(c)
                    except Exception:
                        continue
                    if overlap > best_overlap:
                        best_overlap = overlap
                        chosen_image_col = c
                        norms_for_chosen = norms

            if chosen_image_col is None:
                continue

            print(f"✅ REFUGE_VAL Excel: sheet '{sheet}' → using image id column '{chosen_image_col}' (overlap with IMG/: {best_overlap})")
            # --- NEW DEBUG: show a few sample entries from the chosen column
            sample_vals = df[chosen_image_col].dropna().astype(str).head(10).tolist()
            print(f"   [DBG] First 10 IDs in '{chosen_image_col}': {sample_vals}")

            lbl_col = cols_lower[label_col_lower]
            unknowns = 0
            for _, row in df.iterrows():
                img_id = row.get(chosen_image_col, None)
                if pd.isna(img_id):
                    continue
                lab = self._parse_binary_label(row.get(lbl_col, None))
                if lab is None:
                    unknowns += 1
                    continue
                key = self._norm_stem(img_id)
                label_map[key] = int(lab)

            print(f"   → built {len(label_map)} labeled entries from '{sheet}' (unknowns skipped: {unknowns}).")
            # --- NEW DEBUG: show a few keys we actually built
            dbg_keys = list(label_map.keys())[:10]
            print(f"   [DBG] First 10 keys in label_map: {dbg_keys}")
            found = True
            break

        if not found:
            raise ValueError(
                f"Could not find a sheet in '{xlsx_path}' with '{label_col_name}' and a usable image-id column."
            )
        return label_map



    def load_refuge_val(self, labels_xlsx_path: Optional[str] = None):
        """Load REFUGE_VAL with JPG images + BMP masks; auto-discover labels .xlsx in folder."""
        print("📦 Loading REFUGE_VAL...")

        refuge_val_folder = self._first_existing("REFUGE_VAL", "Refuge_Val", "refuge_val")
        if not refuge_val_folder:
            refuge_val_folder = self._find_folder_containing("refuge_val", "refuge-validation", "validation")
        if not refuge_val_folder:
            print("⚠️ REFUGE_VAL folder not found")
            return []

        root = os.path.join(self.dataset_root, refuge_val_folder)
        print(f"📂 REFUGE_VAL folder contents: {os.listdir(root)}")

        img_root = os.path.join(root, "IMG")
        gt_root  = os.path.join(root, "GT")
        if not (os.path.isdir(img_root) and os.path.isdir(gt_root)):
            print("⚠️ REFUGE_VAL IMG or GT folder not found")
            return []

        # Gather JPGs first so we can build known_basenames for Excel matching
        jpgs = []
        for pat in ["*.jpg", "*.jpeg", "*.JPG", "*.JPEG"]:
            jpgs.extend(glob(os.path.join(img_root, pat)))

        known_basenames = { self._norm_stem(os.path.splitext(os.path.basename(p))[0]) for p in jpgs }

        # ---- Auto-discover Excel label file if not provided
        label_map = {}
        if labels_xlsx_path is None:
            # search only the REFUGE_VAL root for Excel files
            xlsx_candidates = glob(os.path.join(root, "*.xlsx")) + glob(os.path.join(root, "*.xls"))
            picked = None
            for x in xlsx_candidates:
                try:
                    _ = self._load_label_map_from_excel(
                        x, label_col_name="Glaucoma Label", known_img_basenames=known_basenames
                    )
                    picked = x
                    break
                except Exception:
                    continue

            if picked is not None:
                print(f"🔎 Using REFUGE_VAL labels Excel: {os.path.basename(picked)}")
                try:
                    label_map = self._load_label_map_from_excel(
                        picked, label_col_name="Glaucoma Label", known_img_basenames=known_basenames
                    )
                except Exception as e:
                    print(f"⚠️ Failed reading REFUGE_VAL Excel ({picked}): {e}")
                    label_map = {}
            else:
                print("⚠️ No suitable Excel with 'Glaucoma Label' found in REFUGE_VAL folder. Proceeding with label=-1.")
                label_map = {}
        else:
            # explicit path still supported
            try:
                label_map = self._load_label_map_from_excel(
                    labels_xlsx_path, label_col_name="Glaucoma Label", known_img_basenames=known_basenames
                )
            except Exception as e:
                print(f"⚠️ Failed reading REFUGE_VAL Excel ({labels_xlsx_path}): {e}")
                label_map = {}

        # === NEW: expand numeric keys once, to cover variants like V0333, 0333, val0333, etc.
        label_map = self._expand_numeric_keys(label_map)

        def _find_mask(base):
            for ext in [".bmp", ".BMP"]:
                p = os.path.join(gt_root, base + ext)
                if os.path.exists(p):
                    return p
            return None

        samples = []
        total, paired, labeled, unlabeled = 0, 0, 0, 0
        for img_path in jpgs:
            total += 1
            base = os.path.splitext(os.path.basename(img_path))[0]
            mask_path = _find_mask(base)
            if not mask_path:
                continue

            stem = self._norm_stem(base)

            # Simple lookup thanks to expanded keys
            label = label_map.get(stem, None)
            if label is None:
                # optional: trim scan/index suffixes (e.g., v0333_scan1 → v0333)
                key2 = re.sub(r'(?:[-_]?scan)?\d+$', '', stem)
                if key2 and key2 != stem:
                    label = label_map.get(key2, None)

            if label is None:
                label = -1; unlabeled += 1
            else:
                labeled += 1

            samples.append({
                "image": img_path,
                "mask": mask_path,
                "label": int(label),
                "dataset": "REFUGE_VAL"
            })
            paired += 1

        print(f"✅ REFUGE_VAL: {paired} paired images (labels: {labeled} found, {unlabeled} missing) out of {total} files.")
        return samples





    def load_all_datasets(self, include_chaksu_labels: Optional[Dict[str, int]] = None):
        """Load and combine all available datasets"""
        print("="*60)
        print("Loading all available datasets...")
        print("="*60)

        all_samples = []

        # Load each dataset
        datasets = [
            ("Drishti-GS", self.load_drishti_gs_disc_only()),
            ("RIM-ONE-DL", self.load_rimone_dl_disc_only()),
            ("G1020", self.load_g1020_split()),
            ("ORIGA", self.load_origa_split()),
            ("REFUGE", self.load_refuge()),
            ("REFUGE_VAL", self.load_refuge_val()),
            ("Chaksu", self.load_chaksu_img_gt())
        ]

        for name, samples in datasets:
            if samples:
                all_samples.extend(samples)
                print(f"Added {len(samples)} samples from {name}")


        # Filter out samples without valid labels
        valid_samples = [s for s in all_samples if s['label'] in [0, 1]]

        print("\n" + "="*60)
        print(f"✅ Total valid samples loaded: {len(valid_samples)}")

        # Print dataset statistics
        dataset_stats = {}
        for sample in valid_samples:
            ds = sample['dataset']
            if ds not in dataset_stats:
                dataset_stats[ds] = {'glaucoma': 0, 'normal': 0}
            if sample['label'] == 1:
                dataset_stats[ds]['glaucoma'] += 1
            else:
                dataset_stats[ds]['normal'] += 1

        print("\nDataset Statistics:")
        print("-"*40)
        for ds, stats in dataset_stats.items():
            total = stats['glaucoma'] + stats['normal']
            print(f"{ds:15} | Total: {total:4} | Glaucoma: {stats['glaucoma']:3} | Normal: {stats['normal']:3}")

        print("="*60)


        # After printing dataset statistics
        # =================================

        # --- Create label-based folders for Chaksu with images & masks ---
        # --- Combined split for ALL datasets into Fundus_results/glaucoma + control ---
        #output_root = os.path.join(self.results_dir, "combined_split")

        # Delete old split to start fresh
        #if os.path.exists(output_root):
        #    shutil.rmtree(output_root)

        #for label_name in ["glaucoma", "control"]:
        #    os.makedirs(os.path.join(output_root, label_name, "images"), exist_ok=True)
        #    os.makedirs(os.path.join(output_root, label_name, "masks"), exist_ok=True)

        #for sample in tqdm(valid_samples, desc="Creating combined split (ALL datasets)"):
        #    label_dir = "glaucoma" if sample['label'] == 1 else "control"

            # Image
         #   src_img = sample['image']
         #   dst_img = os.path.join(output_root, label_dir, "images", os.path.basename(src_img))
         #   if not os.path.exists(dst_img):
         #       try:
         #           os.symlink(src_img, dst_img)
         #       except OSError:
         #           shutil.copy2(src_img, dst_img)

            # Mask (if present)
         #   src_mask = sample.get('mask', None)
         #   if src_mask and os.path.exists(src_mask):
         #       dst_mask = os.path.join(output_root, label_dir, "masks", os.path.basename(src_mask))
         #       if not os.path.exists(dst_mask):
         #           try:
         #               os.symlink(src_mask, dst_mask)
         #           except OSError:
         #               shutil.copy2(src_mask, dst_mask)

        #print(f"✅ Combined ALL datasets into: {output_root}")
        self.all_samples = valid_samples
        return valid_samples

        # # --- 2) NEW: Split ALL datasets (images + masks) into glaucoma/control ---
        # all_split_root = os.path.join(self.dataset_root, "all_datasets_split")

        # # Clean old to avoid leftovers (optional, comment out if you prefer to keep)
        # if os.path.exists(all_split_root):
        #     shutil.rmtree(all_split_root)

        # for label_name in ["glaucoma", "control"]:
        #     os.makedirs(os.path.join(all_split_root, label_name, "images"), exist_ok=True)
        #     os.makedirs(os.path.join(all_split_root, label_name, "masks"), exist_ok=True)

        # for sample in valid_samples:
        #     label_dir = "glaucoma" if sample['label'] == 1 else "control"

        #     # link/copy image
        #     src_img = sample['image']
        #     dst_img = os.path.join(all_split_root, label_dir, "images", os.path.basename(src_img))
        #     if not os.path.exists(dst_img):
        #         try:
        #             os.symlink(src_img, dst_img)
        #         except OSError:
        #             shutil.copy2(src_img, dst_img)

        #     # link/copy mask (if present)
        #     src_mask = sample.get('mask', None)
        #     if src_mask is not None and os.path.exists(src_mask):
        #         dst_mask = os.path.join(all_split_root, label_dir, "masks", os.path.basename(src_mask))
        #         if not os.path.exists(dst_mask):
        #             try:
        #                 os.symlink(src_mask, dst_mask)
        #             except OSError:
        #                 shutil.copy2(src_mask, dst_mask)

        # print(f"✅ Split for ALL datasets (images + masks) saved to: {all_split_root}")
        # self.all_samples = valid_samples
        # return valid_samples
    def prepare_for_training(self, samples: List[Dict], vcdr_values: Optional[Dict] = None):
        """Convert samples to format expected by the pipeline"""
        image_paths = []
        labels = []
        masks = []
        vcdr_list = []

        for sample in samples:
            image_paths.append(sample['image'])
            labels.append(sample['label'])
            masks.append(sample['mask'])

            # Add VCDR if available
            if vcdr_values and sample['image'] in vcdr_values:
                vcdr_list.append(vcdr_values[sample['image']])
            else:
                vcdr_list.append(np.nan)

        return image_paths, labels, masks, vcdr_list

# =============================================================================
# MODIFIED OPTIC DISC SEGMENTER TO USE PRE-EXISTING MASKS
# =============================================================================

class OpticDiscSegmenter:
    """Handles optic disc segmentation using pre-existing masks or training new model"""

    def __init__(self, model_path: Optional[str] = None, use_pretrained_masks: bool = True):
        self.use_pretrained_masks = use_pretrained_masks
        self.mask_cache = {}  # Cache for pre-loaded masks

        self.model = OpticDiscSegmentationModel().to(device)
        self.criterion = DiceCELoss(sigmoid=True)

        if model_path and os.path.exists(model_path):
            self.model.load_state_dict(torch.load(model_path, map_location=device))
            print(f"Loaded pretrained model from {model_path}")

        # Inference transform (image only)
        self.infer_transform = Compose([
            LoadImaged(keys=["image"]),
            EnsureChannelFirstd(keys=["image"]),
            ScaleIntensityRanged(keys=["image"], a_min=0, a_max=255, b_min=0.0, b_max=1.0, clip=True),
            Resized(keys=["image"], spatial_size=(512, 512)),
            EnsureTyped(keys=["image"]),
        ])

        # Training transform (image + label)
        self.train_transform = Compose([
            LoadImaged(keys=["image", "label"]),
            EnsureChannelFirstd(keys=["image", "label"]),
            ScaleIntensityRanged(keys=["image"], a_min=0, a_max=255, b_min=0.0, b_max=1.0, clip=True),
            ScaleIntensityRanged(keys=["label"], a_min=0, a_max=255, b_min=0.0, b_max=1.0, clip=True),  # <-- scale labels too
            Resized(keys=["image", "label"], spatial_size=(512, 512)),
            EnsureTyped(keys=["image", "label"]),
            AsDiscreted(keys="label", threshold=0.5),
        ])

        self.postprocess = Compose([
            EnsureTyped(keys="pred"),
            Activationsd(keys="pred", sigmoid=True),
            AsDiscreted(keys="pred", threshold=0.5),
        ])

    def load_mask_mappings(self, samples: List[Dict]):
        """Pre-load mask mappings for faster access"""
        print("Loading mask mappings...")
        for sample in samples:
            self.mask_cache[sample['image']] = sample['mask']
        print(f"Loaded {len(self.mask_cache)} mask mappings")

    def segment_optic_disc(self, image_path: str) -> Tuple[np.ndarray, Tuple[int, int]]:
        """Segment optic disc using pre-existing mask or model"""

        # First, try to use pre-existing mask (robust to format & polarity)
        if self.use_pretrained_masks and image_path in self.mask_cache:
            mask_path = self.mask_cache[image_path]
            if os.path.exists(mask_path):
                mask = cv2.imread(mask_path, cv2.IMREAD_UNCHANGED)
                if mask is not None:
                    # to grayscale
                    if mask.ndim == 3:
                        mask = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)

                    # normalize possible {0,1} to {0,255}
                    if mask.max() <= 1:
                        mask = (mask.astype(np.float32) * 255).astype(np.uint8)

                    # Otsu threshold
                    _, th = cv2.threshold(mask, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
                    bin1 = (th > 0).astype(np.uint8)
                    bin2 = (1 - bin1).astype(np.uint8)  # opposite polarity

                    def plausible(b):
                        frac = float(b.sum()) / b.size
                        return 0.003 <= frac <= 0.10  # disc typically ~0.1–5% of 512x512

                    binm = bin1 if plausible(bin1) else bin2

                    # keep largest connected component
                    num, labels = cv2.connectedComponents(binm)
                    if num > 1:
                        areas = [(labels == i).sum() for i in range(1, num)]
                        comp = 1 + int(np.argmax(areas))
                        binm = (labels == comp).astype(np.uint8)

                    binm = cv2.resize(binm, (512, 512), interpolation=cv2.INTER_NEAREST)
                    centroid = self.compute_centroid(binm)
                    return binm, centroid

        # Fall back to model prediction
        data = {"image": image_path}
        data = self.infer_transform(data)

        self.model.eval()
        with torch.no_grad():
            input_tensor = data["image"].unsqueeze(0).to(device)
            output = sliding_window_inference(input_tensor, (512, 512), 4, self.model)
            output = torch.sigmoid(output).cpu().numpy()

        mask = (output[0, 0] > 0.5).astype(np.uint8)
        centroid = self.compute_centroid(mask)

        return mask, centroid

    def compute_centroid(self, mask: np.ndarray) -> Tuple[int, int]:
        mask = (mask > 0).astype(np.uint8)
        m = cv2.moments(mask)
        if m["m00"] > 0:
            cx = int(m["m10"] / m["m00"])
            cy = int(m["m01"] / m["m00"])
            return cx, cy
        h, w = mask.shape
        return w // 2, h // 2

    def train_segmentation_model(self, train_samples: List[Dict], val_samples: List[Dict],
                                num_epochs: int = 50, learning_rate: float = 5e-5):
        """Train the optic disc segmentation model using provided masks"""

        # Prepare data for MONAI
        train_data = []
        for sample in train_samples:
            train_data.append({
                "image": sample["image"],
                "label": sample["mask"]
            })

        val_data = []
        for sample in val_samples:
            val_data.append({
                "image": sample["image"],
                "label": sample["mask"]
            })

        # Create MONAI datasets
        train_ds = MonaiDataset(data=train_data, transform=self.train_transform)
        val_ds = MonaiDataset(data=val_data, transform=self.train_transform)

        train_loader = MonaiDataLoader(train_ds, batch_size=4, shuffle=True, num_workers=2)
        val_loader = MonaiDataLoader(val_ds, batch_size=1, num_workers=2)

        optimizer = optim.Adam(self.model.parameters(), lr=learning_rate)
        scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=5, verbose=True)


        dice_metric = DiceMetric(include_background=False, reduction="mean")

        best_metric = -1
        best_metric_epoch = -1

        for epoch in trange(num_epochs, desc="[seg] epochs", leave=True):
            print(f"Epoch {epoch + 1}/{num_epochs}")
            self.model.train()
            epoch_loss = 0
            step = 0

            for batch_data in train_loader:
                step += 1
                inputs, labels = batch_data["image"].to(device), batch_data["label"].to(device)
                optimizer.zero_grad()
                outputs = self.model(inputs)
                loss = self.criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                epoch_loss += loss.item()

            print(f"Epoch {epoch + 1} average loss: {epoch_loss / step:.4f}")

            # Validation every 10 epochs
            self.model.eval()
            with torch.no_grad():
                for val_data in val_loader:
                    val_inputs = val_data["image"].to(device)
                    val_outputs = sliding_window_inference(val_inputs, (512, 512), 4, self.model)
                    val_outputs = [self.postprocess({"pred": i})["pred"] for i in decollate_batch(val_outputs)]
                    val_labels = [val_data["label"][i] for i in range(len(val_outputs))]
                    dice_metric(y_pred=val_outputs, y=val_labels)

                metric = dice_metric.aggregate().item()
                dice_metric.reset()

                if metric > best_metric:
                    best_metric = metric
                    best_metric_epoch = epoch + 1
                    torch.save(self.model.state_dict(), "best_optic_disc_model.pth")
                    print(f"Saved new best metric model at epoch {best_metric_epoch}, metric: {best_metric:.4f}")

            scheduler.step()

        print(f"Training completed, best metric: {best_metric:.4f} at epoch: {best_metric_epoch}")

# Keep all other original classes (OpticDiscSegmentationModel, ImageCropper, etc.)
class OpticDiscSegmentationModel(nn.Module):
    """U-Net model for optic disc segmentation using MONAI"""

    def __init__(self, spatial_dims=2, in_channels=3, out_channels=1, channels=(16, 32, 64, 128, 256), strides=(2, 2, 2, 2)):
        super().__init__()
        self.unet = UNet(
            spatial_dims=spatial_dims,
            in_channels=in_channels,
            out_channels=out_channels,
            channels=channels,
            strides=strides,
            num_res_units=2,
            norm=Norm.BATCH,
        )

    def forward(self, x):
        return self.unet(x)

class ImageCropper:
    """Handles ONH and periphery cropping policies"""

    @staticmethod
    def create_circular_mask(image_shape: Tuple[int, int], center: Tuple[int, int],
                           radius: int, inverse: bool = False) -> np.ndarray:
        """Create circular mask"""
        h, w = image_shape
        y, x = np.ogrid[:h, :w]
        mask = (x - center[0])**2 + (y - center[1])**2 <= radius**2
        return ~mask if inverse else mask

    def onh_crop_policy(self, image: np.ndarray, centroid: Tuple[int, int],
                      mask_radius_percent: float) -> Tuple[np.ndarray, np.ndarray]:
        """ONH Crop Policy: Remove ONH spatial information using inpainting"""
        h, w = image.shape[:2]
        image_size = min(h, w)
        radius = int((mask_radius_percent / 100.0) * image_size / 2)

        mask = self.create_circular_mask((h, w), centroid, radius)
        mask_uint8 = mask.astype(np.uint8) * 255

        # Inpaint to remove spatial cues
        result = cv2.inpaint(image, mask_uint8, inpaintRadius=radius//3, flags=cv2.INPAINT_TELEA)

        return result, mask

    def periphery_crop_policy(self, image: np.ndarray, centroid: Tuple[int, int],
                            keep_radius_percent: float) -> np.ndarray:
        """Periphery Crop Policy: Keep only optic disc, mask everything outside"""
        h, w = image.shape[:2]
        image_size = min(h, w)
        radius = int((keep_radius_percent / 100.0) * image_size / 2)

        mask = self.create_circular_mask((h, w), centroid, radius, inverse=True)
        result = image.copy()
        result[mask] = 0

        return result

class GlaucomaClassificationModel(nn.Module):
    """ResNet-50 based glaucoma detection model"""

    def __init__(self, num_classes: int = 1, include_vcdr: bool = False, reduce_spatial_bias: bool = False):
        super().__init__()
        self.backbone = resnet50(weights=ResNet50_Weights.IMAGENET1K_V1)
        self.backbone = nn.Sequential(*list(self.backbone.children())[:-1])

        if reduce_spatial_bias:
            self.spatial_dropout = nn.Dropout2d(p=0.3)
        else:
            self.spatial_dropout = nn.Identity()

        self.global_avg_pool = nn.AdaptiveAvgPool2d((1, 1))

        self.classifier = nn.Sequential(
            nn.Dropout(0.5),
            nn.Linear(2048, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, num_classes)
        )

        self.include_vcdr = include_vcdr
        if include_vcdr:
            self.vcdr_regressor = nn.Sequential(
                nn.Dropout(0.5),
                nn.Linear(2048, 1)
            )

    def forward(self, x):
        features = self.backbone(x)
        features = self.spatial_dropout(features)
        features = self.global_avg_pool(features)
        features = torch.flatten(features, 1)
        classification = self.classifier(features)


        if self.include_vcdr:
            vcdr = self.vcdr_regressor(features)
            return classification, vcdr

        return classification

# =============================
# Grad-CAM for ResNet-50 head
# =============================
class GradCAM:
    def __init__(self, clf_model: GlaucomaClassificationModel, device: torch.device, custom_transform=None):
        self.model = clf_model
        self.device = device
        self.model.eval()

        # target conv layer: resnet layer4  (index 7 in our backbone Sequential)
        self.target_layer = self.model.backbone[7]

        self.activations = None
        self.gradients = None

        def fwd_hook(module, inp, out):
            self.activations = out.detach()

        def bwd_hook(module, grad_in, grad_out):
            self.gradients = grad_out[0].detach()

        self.fwd_handle = self.target_layer.register_forward_hook(fwd_hook)
        self.bwd_handle = self.target_layer.register_full_backward_hook(bwd_hook)

        # Use custom transform if provided, otherwise use default
        if custom_transform is not None:
            self.preprocess = custom_transform
        else:
            self.preprocess = transforms.Compose([
                transforms.Resize((512, 512)),
                transforms.ToTensor(),
                transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                     std=[0.229, 0.224, 0.225])
            ])

    def __del__(self):
        try:
            self.fwd_handle.remove()
            self.bwd_handle.remove()
        except Exception:
            pass

    def _compute_cam(self) -> np.ndarray:
        """
        Build CAM from cached activations and gradients.
        Returns HxW numpy map in [0,1].
        """
        # activations: [B, C, H, W]; gradients: [B, C, H, W]
        weights = torch.mean(self.gradients, dim=(2, 3), keepdim=True)  # [B,C,1,1]
        cam = torch.sum(weights * self.activations, dim=1)              # [B,H,W]
        cam = torch.relu(cam)
        # normalize to [0,1]
        cam_min = cam.min(dim=-1, keepdim=True)[0].min(dim=-2, keepdim=True)[0]
        cam_max = cam.max(dim=-1, keepdim=True)[0].max(dim=-2, keepdim=True)[0]
        cam = (cam - cam_min) / (cam_max - cam_min + 1e-8)
        return cam[0].cpu().numpy()

    def _heatmap_on_image(self, heatmap: np.ndarray, rgb_img_512: np.ndarray,
                          true_label: int, pred_prob: float, out_path: str):
        """
        Overlay heatmap on image and save with legend.
        """
        h, w = rgb_img_512.shape[:2]
        if heatmap.shape != (h, w):
            heatmap = cv2.resize(heatmap, (w, h), interpolation=cv2.INTER_CUBIC)


        # NEW: Zero out attention in black/masked regions
        gray_img = cv2.cvtColor(rgb_img_512, cv2.COLOR_RGB2GRAY)
        black_mask = gray_img < 10  # Detect very dark/black pixels
        heatmap[black_mask] = 0     # Force attention to 0 in masked areas

        # Renormalize heatmap after masking
        if heatmap.max() > 0:
            heatmap = heatmap / heatmap.max()

        # Create a threshold for significant attention (e.g., > 0.3)
        attention_threshold = 0.3
        attention_mask = heatmap > attention_threshold

        # colorize heatmap
        heat_uint8 = np.uint8(255 * np.clip(heatmap, 0, 1))
        heat_color = cv2.applyColorMap(heat_uint8, cv2.COLORMAP_JET)[:, :, ::-1]  # BGR->RGB

        # Create overlay: only apply heatmap where attention is significant
        overlay = rgb_img_512.astype(np.uint8).copy()

        # Apply heatmap only to areas with significant attention
        for c in range(3):  # RGB channels
            overlay[:, :, c] = np.where(
                attention_mask,
                (0.35 * heat_color[:, :, c] + 0.65 * rgb_img_512[:, :, c]).astype(np.uint8),
                rgb_img_512[:, :, c]  # Keep original image where attention is low
            )

        # ---- Matplotlib figure with colorbar + text ----
        fig, ax = plt.subplots(figsize=(6, 6))
        im = ax.imshow(overlay)
        ax.axis("off")

        # add colorbar (blue→red attention scale)
        cax = fig.add_axes([0.85, 0.15, 0.04, 0.7])
        cb = plt.colorbar(plt.cm.ScalarMappable(cmap="jet"), cax=cax)
        cb.set_label("Attention (low→high)", rotation=90)

        # add legend text with both true and predicted labels
        true_label_str = "Glaucoma" if true_label == 1 else "Normal"
        pred_label_str = "Glaucoma" if pred_prob >= 0.5 else "Normal"
        ax.set_title(f"True: {true_label_str} | Pred: {pred_label_str} (prob: {pred_prob:.2f})", fontsize=10)

        plt.savefig(out_path, dpi=200, bbox_inches="tight")
        plt.close(fig)

        return overlay

    def run_on_image(self, image_path: str, true_label: int, out_path: str) -> Tuple[np.ndarray, float]:
        try:
            pil = Image.open(image_path).convert('RGB')
        except Exception as e:
            print(f"Error loading image {image_path}: {e}")
            return np.zeros((512, 512)), 0.0

        # Ensure model is in eval mode
        self.model.eval()

        try:
            tens = self.preprocess(pil).unsqueeze(0).to(self.device).requires_grad_(True)
        except Exception as e:
            print(f"Error preprocessing image {image_path}: {e}")
            return np.zeros((512, 512)), 0.0

        # Clear any existing gradients
        self.model.zero_grad()
        if tens.grad is not None:
            tens.grad.zero_()

        try:
            out = self.model(tens)
            if isinstance(out, tuple):
                out = out[0]
            logit = out.squeeze()

            # Backward pass
            logit.backward(retain_graph=False)

            prob = float(torch.sigmoid(logit).detach().cpu().item())
            heat = self._compute_cam()

            if heat is None or heat.size == 0:
                print(f"Warning: Empty heatmap for {image_path}")
                heat = np.zeros((512, 512))

        except Exception as e:
            print(f"Error during forward/backward pass for {image_path}: {e}")
            return np.zeros((512, 512)), 0.0

        try:
            rgb_512 = np.array(pil.resize((512, 512)))
            self._heatmap_on_image(heat, rgb_512, true_label, prob, out_path)
        except Exception as e:
            print(f"Error creating heatmap overlay for {image_path}: {e}")

        return heat, prob


class FundusDataset(Dataset):
    """Custom dataset for fundus images"""

    def __init__(self, image_paths: List[str], labels: List[int],
                 vcdr_values: Optional[List[float]] = None,
                 transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.vcdr_values = vcdr_values
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image = Image.open(self.image_paths[idx]).convert('RGB')
        label = torch.tensor(self.labels[idx], dtype=torch.float32)

        if self.transform:
            image = self.transform(image)

        sample = {'image': image, 'label': label}

        if self.vcdr_values is not None:
            sample['vcdr'] = torch.tensor(self.vcdr_values[idx], dtype=torch.float32)

        return sample

class GlaucomaTrainer:
    """Handles training and evaluation of glaucoma detection models"""

    def __init__(self, include_vcdr: bool = False, save_dir: str = None, reduce_spatial_bias: bool = False):
        self.include_vcdr = include_vcdr
        self.model = GlaucomaClassificationModel(include_vcdr=include_vcdr, reduce_spatial_bias=reduce_spatial_bias).to(device)
        self.save_dir = save_dir if save_dir else "."

        self.classification_loss = nn.BCEWithLogitsLoss()
        self.use_class_weights = True

        self.regression_loss = nn.MSELoss() if include_vcdr else None

        self.train_transform = transforms.Compose([
            transforms.Resize((512, 512)),
            transforms.RandomHorizontalFlip(p=0.5),
            transforms.RandomRotation(degrees=30),
            transforms.RandomAffine(degrees=0, translate=(0.15, 0.15), scale=(0.9, 1.1)),
            transforms.RandomPerspective(distortion_scale=0.2, p=0.5),
            transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.2),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])

        self.val_transform = transforms.Compose([
            transforms.Resize((512, 512)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])

    def train_model(self, train_dataset: FundusDataset, val_dataset: FundusDataset,
                    num_epochs: int = 50, learning_rate: float = 1e-4, batch_size: int = 16,
                    model_name: str = "glaucoma",
                    early_stop_patience: int = 8, min_delta: float = 0.001):
        """Train the glaucoma detection model with per-epoch history & plots."""

        # Detect where IMAGES live (not the save_dir)
        def _on_drive(ds: FundusDataset) -> bool:
            try:
                p0 = ds.image_paths[0]
                return isinstance(p0, str) and p0.startswith("/content/drive")
            except Exception:
                return False

        imgs_on_drive = _on_drive(train_dataset) or _on_drive(val_dataset)

        # Google Drive + multi-worker DataLoader often hangs → use safest settings
        if imgs_on_drive:
            num_workers = 8
            pin_memory  = False
        else:
            num_workers = min(16, max(2, (os.cpu_count() or 4)//2))
            pin_memory  = torch.cuda.is_available()

        loader_kwargs = dict(
            batch_size=batch_size,
            num_workers=num_workers,
            pin_memory=pin_memory,
            persistent_workers=False,
        )
        if num_workers > 0:
            loader_kwargs["prefetch_factor"] = 2

        train_loader = DataLoader(train_dataset, shuffle=True,  **loader_kwargs)
        val_loader   = DataLoader(val_dataset,   shuffle=False, **loader_kwargs)

        print(f"[{model_name}] dataset sizes -> train={len(train_dataset)}, val={len(val_dataset)}", flush=True)
        print(f"[{model_name}] steps per epoch: train={len(train_loader)}, val={len(val_loader)}", flush=True)
        print(f"[{model_name}] imgs_on_drive={imgs_on_drive}, num_workers={num_workers}", flush=True)

        # Sanity pull (surfaces errors early)
        try:
            _ = next(iter(DataLoader(train_dataset, batch_size=4, num_workers=0)))
            print(f"[{model_name}] single-batch sanity check OK", flush=True)
        except Exception as e:
            print(f"[{model_name}] single-batch sanity check FAILED: {e}", flush=True)
            raise

        optimizer = optim.AdamW(self.model.parameters(), lr=1e-4, weight_decay=1e-2)


        scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=num_epochs)



        best_loss = float("inf")
        no_improve = 0
        model_path = os.path.join(self.save_dir, f'best_{model_name}_model.pth')

        # ---- NEW: training history ----
        history = {
            'epoch': [],
            'train_loss': [],
            'train_acc': [],
            'val_auc': [],
            'val_acc': []
        }

        for epoch in trange(num_epochs, desc=f"[{model_name}] epochs", leave=True):
            print(f"[{model_name}] Epoch {epoch + 1}/{num_epochs}", flush=True)
            self.model.train()
            train_loss = 0.0

            # ---- NEW: compute training accuracy on-the-fly ----
            train_correct = 0
            train_total   = 0

            train_bar = tqdm(
                train_loader,
                total=len(train_loader),
                desc=f"train {epoch+1}/{num_epochs}",
                leave=False,
                position=1,
            )
            for i, batch in enumerate(train_bar, 1):
                images = batch['image'].to(device)
                labels = batch['label'].to(device)

                optimizer.zero_grad()
                if self.include_vcdr:
                    class_logits, vcdr_output = self.model(images)
                    vcdr_targets = batch['vcdr'].to(device)
                    # Apply class weights to balance the loss
                    pos_weight = torch.tensor([3.3], device=device)  # 3686/1113 ≈ 3.3
                    weighted_loss = nn.BCEWithLogitsLoss(pos_weight=pos_weight)
                    class_loss   = weighted_loss(class_logits.squeeze(1), labels)
                    vcdr_loss    = self.regression_loss(vcdr_output.squeeze(1), vcdr_targets)
                    loss = class_loss + vcdr_loss
                    logits_for_acc = class_logits
                else:
                    class_logits = self.model(images)
                    # Apply class weights to balance the loss
                    pos_weight = torch.tensor([3.3], device=device)  # 3686/1113 ≈ 3.3
                    weighted_loss = nn.BCEWithLogitsLoss(pos_weight=pos_weight)
                    loss = weighted_loss(class_logits.squeeze(1), labels)
                    logits_for_acc = class_logits

                loss.backward()
                optimizer.step()
                train_loss += loss.item()

                # accumulate training accuracy
                with torch.no_grad():
                    probs  = torch.sigmoid(logits_for_acc.squeeze(1))
                    preds  = (probs >= 0.5).float()
                    train_correct += (preds == labels).sum().item()
                    train_total   += labels.numel()

                # live running average on the bar
                train_bar.set_postfix(loss=f"{train_loss/i:.4f}")

            # end epoch: aggregate
            avg_train_loss = train_loss / max(1, len(train_loader))
            train_acc      = (train_correct / max(1, train_total)) if train_total > 0 else float('nan')

            # Validation
            # Validation
            val_loss = 0.0
            val_correct = 0
            val_total = 0

            self.model.eval()
            with torch.no_grad():
                val_bar = tqdm(val_loader, desc="val", leave=False, position=1)
                for batch in val_bar:
                    images = batch['image'].to(device)
                    labels = batch['label'].to(device)

                    if self.include_vcdr:
                        class_logits, vcdr_output = self.model(images)
                        vcdr_targets = batch['vcdr'].to(device)
                        pos_weight = torch.tensor([3.3], device=device)
                        weighted_loss = nn.BCEWithLogitsLoss(pos_weight=pos_weight)
                        class_loss = weighted_loss(class_logits.squeeze(1), labels)
                        vcdr_loss = self.regression_loss(vcdr_output.squeeze(1), vcdr_targets)
                        loss = class_loss + vcdr_loss
                        logits_for_acc = class_logits
                    else:
                        class_logits = self.model(images)
                        pos_weight = torch.tensor([3.3], device=device)
                        weighted_loss = nn.BCEWithLogitsLoss(pos_weight=pos_weight)
                        loss = weighted_loss(class_logits.squeeze(1), labels)
                        logits_for_acc = class_logits

                    val_loss += loss.item()

                    # Calculate validation accuracy
                    probs = torch.sigmoid(logits_for_acc.squeeze(1))
                    preds = (probs >= 0.5).float()
                    val_correct += (preds == labels).sum().item()
                    val_total += labels.numel()

                    val_bar.set_postfix(loss=f"{val_loss/len(val_loader):.4f}")

            avg_val_loss = val_loss / max(1, len(val_loader))
            val_acc = (val_correct / max(1, val_total)) if val_total > 0 else float('nan')

            # Calculate AUC for reporting (but not for model selection)
            val_metrics = self.evaluate_model(val_loader, show_bar=False, desc="val-auc")
            val_auc = float(val_metrics.get('auc', float('nan')))

            tqdm.write(f"Epoch {epoch+1}/{num_epochs}")
            tqdm.write(f"Train Loss: {avg_train_loss:.4f} | Train Acc: {train_acc:.4f}")
            tqdm.write(f"Val Loss  : {avg_val_loss:.4f}  | Val AUC  : {val_auc:.4f}")

            # ---- NEW: record history ----
            history['epoch'].append(int(epoch + 1))
            history['train_loss'].append(float(avg_train_loss))
            history['train_acc'].append(float(train_acc))
            history['val_loss'] = history.get('val_loss', [])  # Add val_loss to history
            history['val_loss'].append(float(avg_val_loss))
            history['val_auc'].append(float(val_auc))
            history['val_acc'].append(float(val_acc))

            # Model selection by validation loss (lower is better)
            current_val_loss = avg_val_loss
            if math.isfinite(current_val_loss) and (best_loss - current_val_loss) > min_delta:
                best_loss = current_val_loss
                no_improve = 0
                torch.save(self.model.state_dict(), model_path)
                print(f"New best model saved to {model_path} with val loss: {best_loss:.4f}")
            else:
                no_improve += 1
                print(f"No improvement ({no_improve}/{early_stop_patience})")

            # early stopping
            if no_improve >= early_stop_patience:
                print(f"⏹️ Early stopping at epoch {epoch+1}: no AUC improvement > {min_delta} for {early_stop_patience} epochs.")
                break

            scheduler.step()

            print("-" * 50)

        # ---- NEW: save curves + raw history (PNG + CSV + JSON) ----
        try:
            import matplotlib.pyplot as plt

            # Accuracy curve
            plt.figure(figsize=(8, 6))
            plt.plot(history['epoch'], history['train_acc'], label="Train Accuracy")
            plt.plot(history['epoch'], history['val_acc'],   label="Validation Accuracy")
            plt.xlabel("Epoch"); plt.ylabel("Accuracy")
            plt.title(f"Accuracy Curve ({model_name})")
            plt.grid(True, alpha=0.3); plt.legend()
            acc_png = os.path.join(self.save_dir, f"{model_name}_acc_curve.png")
            plt.savefig(acc_png, dpi=300, bbox_inches="tight"); plt.close()
            print(f"[{model_name}] Saved: {acc_png}")

            # Loss curves (train vs validation)
            plt.figure(figsize=(8, 6))
            plt.plot(history['epoch'], history['train_loss'], label="Train Loss")
            plt.plot(history['epoch'], history['val_loss'], label="Validation Loss")
            plt.xlabel("Epoch"); plt.ylabel("Loss")
            plt.title(f"Loss Curves ({model_name})")
            plt.grid(True, alpha=0.3); plt.legend()
            loss_png = os.path.join(self.save_dir, f"{model_name}_loss_curves.png")
            plt.savefig(loss_png, dpi=300, bbox_inches="tight"); plt.close()
            print(f"[{model_name}] Saved: {loss_png}")
        except Exception as e:
            print(f"({model_name}) Plotting skipped: {e}")

        # CSV + JSON logs
        try:
            csv_path = os.path.join(self.save_dir, f"{model_name}_training_history.csv")
            with open(csv_path, "w", newline="") as f:
                w = csv.writer(f)
                w.writerow(["epoch", "train_loss", "train_acc", "val_auc", "val_acc"])
                for i in range(len(history['epoch'])):
                    w.writerow([
                        history['epoch'][i],
                        history['train_loss'][i],
                        history['train_acc'][i],
                        history['val_auc'][i],
                        history['val_acc'][i],
                    ])
            print(f"[{model_name}] Saved: {csv_path}")

            json_path = os.path.join(self.save_dir, f"{model_name}_training_history.json")
            with open(json_path, "w") as f:
                json.dump(history, f, indent=2)
            print(f"[{model_name}] Saved: {json_path}")
        except Exception as e:
            print(f"[{model_name}] Failed to save history logs: {e}")


    def evaluate_model(self, data_loader: DataLoader, show_bar: bool = True, desc: str = "eval") -> Dict[str, float]:
        self.model.eval()
        all_predictions, all_labels = [], []
        all_vcdr_pred, all_vcdr_true = [], []

        with torch.no_grad():
            iterator = tqdm(
                data_loader,
                total=len(data_loader),
                desc=desc,
                leave=False,
                unit="batch"
            ) if show_bar else data_loader

            for batch in iterator:
                images = batch['image'].to(device)
                labels = batch['label'].cpu().numpy()

                if self.include_vcdr:
                    class_logits, vcdr_output = self.model(images)
                    vcdr_pred = vcdr_output.cpu().numpy()
                    vcdr_true = batch['vcdr'].cpu().numpy()
                    all_vcdr_pred.extend(vcdr_pred.flatten())
                    all_vcdr_true.extend(vcdr_true.flatten())
                else:
                    class_logits = self.model(images)

                logits = class_logits.cpu().numpy()
                all_predictions.extend(logits.flatten())
                all_labels.extend(labels.flatten())

                if show_bar and hasattr(iterator, "set_postfix"):
                    iterator.set_postfix(samples=len(all_labels))

        metrics = {}

        # --- AUC (guard if only one class is present)
        if len(np.unique(all_labels)) < 2:
            metrics['auc'] = float('nan')
        else:
            try:
                metrics['auc'] = roc_auc_score(all_labels, all_predictions)
            except ValueError:
                metrics['auc'] = float('nan')

        # --- Thresholded metrics
        probs = 1.0 / (1.0 + np.exp(-np.array(all_predictions)))     # sigmoid
        pred_labels = (probs >= 0.5).astype(int)
        y_true = np.array(all_labels, dtype=int)

        # Accuracy
        try:
            metrics['accuracy'] = accuracy_score(y_true, pred_labels)
        except Exception:
            metrics['accuracy'] = float('nan')

        # Precision / Recall / F1 (binary; avoid ZeroDivision errors)
        try:
            metrics['precision'] = precision_score(y_true, pred_labels, zero_division=0)
            metrics['recall']    = recall_score(y_true, pred_labels, zero_division=0)
            metrics['f1']        = f1_score(y_true, pred_labels, zero_division=0)
        except Exception:
            metrics['precision'] = float('nan')
            metrics['recall']    = float('nan')
            metrics['f1']        = float('nan')

        # Optional: VCDR regression metrics if enabled
        if self.include_vcdr and len(all_vcdr_pred) > 0:
            r2 = r2_score(all_vcdr_true, all_vcdr_pred)
            mae = mean_absolute_error(all_vcdr_true, all_vcdr_pred)
            pearson_r, _ = pearsonr(all_vcdr_true, all_vcdr_pred)
            metrics.update({'r2': r2, 'mae': mae, 'pearson_r': pearson_r})

        return metrics



    # ----------------------
    # NEW: Grad-CAM helpers
    # ----------------------
    def generate_and_save_gradcams(self, image_paths: List[str], out_dir: str,
                                  max_images: int = 50,
                                  labels: List[int] = None):
        """
        Runs Grad-CAM on up to `max_images` images and saves results
        """
        os.makedirs(out_dir, exist_ok=True)

        # Use consistent transforms by passing val_transform to GradCAM
        cammer = GradCAM(self.model, device, custom_transform=self.val_transform)

        if labels is None:
            raise ValueError("labels must be provided so the legend can show the true class.")

        count = 0
        failed = 0
        for i, p in enumerate(image_paths):
            if count >= max_images:
                break

            try:
                base = os.path.splitext(os.path.basename(p))[0]
                out_img = os.path.join(out_dir, f"{base}_overlay_legend.jpg")

                heat, prob = cammer.run_on_image(p, int(labels[i]), out_img)

                # Save heatmap and score
                np.save(os.path.join(out_dir, f"{base}_heat.npy"), heat)
                with open(os.path.join(out_dir, f"{base}_score.txt"), "w") as f:
                    f.write(f"sigmoid_prob={prob:.6f}\n")

                count += 1

            except Exception as e:
                failed += 1
                print(f"[GradCAM] Failed {p}: {e}")
                if failed > 10:  # Stop if too many failures
                    print("Too many Grad-CAM failures, stopping")
                    break

        print(f"Generated {count} Grad-CAM visualizations, {failed} failed")

# =============================================================================
# MODIFIED GLAUCOMA PIPELINE TO USE MULTI-DATASET LOADER
# =============================================================================

class GlaucomaPipeline:
    """Main pipeline orchestrator with multi-dataset support"""

    def __init__(self, segmentation_model_path: Optional[str] = None,
                 use_pretrained_masks: bool = True,
                 results_dir: str = "/content/drive/MyDrive/Fundus_results"):
        self.segmenter = OpticDiscSegmenter(segmentation_model_path, use_pretrained_masks)
        self.cropper = ImageCropper()
        self.results = {}
        self.results_dir = results_dir

        # Create results directory if it doesn't exist
        os.makedirs(self.results_dir, exist_ok=True)

    def process_dataset_with_masks(self, samples: List[Dict],
                                  crop_percentages: List[float] = [10, 20, 30, 40, 50, 60]):
        """Process dataset with pre-existing masks"""

        print("Starting dataset processing with pre-existing masks...")

        # Load mask mappings for faster access
        self.segmenter.load_mask_mappings(samples)


        # Create directories for cropped images in results directory
        cropped_dir = os.path.join(self.results_dir, "cropped_images")
        os.makedirs(os.path.join(cropped_dir, "full"), exist_ok=True)
        os.makedirs(os.path.join(cropped_dir, "onh_crop"), exist_ok=True)
        os.makedirs(os.path.join(cropped_dir, "periphery_crop"), exist_ok=True)

        # OPTIONAL: create subfolders per percentage for tidier browsing
        for pct in crop_percentages:
            os.makedirs(os.path.join(cropped_dir, "onh_crop", f"{int(pct)}pct"), exist_ok=True)
            os.makedirs(os.path.join(cropped_dir, "periphery_crop", f"{int(pct)}pct"), exist_ok=True)


        processed_data = { 'full': {'paths': [], 'labels': [], 'vcdr': []},
                          'onh_crop': {pct: {'paths': [], 'labels': [], 'vcdr': []}
                                      for pct in crop_percentages},
                          'periphery_crop': {pct: {'paths': [], 'labels': [], 'vcdr': []}
                                            for pct in crop_percentages} }

        kept_to_original = []
        for i, sample in enumerate(samples):
            label = sample['label']
            if i % 50 == 0:
                print(f"Processing image {i+1}/{len(samples)}...")

            image_path = sample['image']

            # Load original image
            image = cv2.imread(image_path)
            if image is None:
                print(f"Warning: Could not load image {image_path}")
                continue

            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            image = cv2.resize(image, (512, 512))

            # Get mask and centroid
            mask, centroid = self.segmenter.segment_optic_disc(image_path)
            vcdr = float('nan')


            # Save full image
            idx_out = len(processed_data['full']['paths'])
            kept_to_original.append(i)
            full_path = os.path.join(cropped_dir, "full", f"image_{idx_out:04d}.jpg")
            cv2.imwrite(full_path, cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
            processed_data['full']['paths'].append(full_path)
            processed_data['full']['labels'].append(label)
            processed_data['full']['vcdr'].append(vcdr)

            # Process with different crop percentages
            for pct in crop_percentages:
                # ONH crop policy
                onh_cropped, onh_mask = self.cropper.onh_crop_policy(image, centroid, pct)
                onh_path = os.path.join(cropped_dir, "onh_crop", f"{int(pct)}pct", f"image_{idx_out:04d}.jpg")

                cv2.imwrite(onh_path, cv2.cvtColor(onh_cropped, cv2.COLOR_RGB2BGR))
                processed_data['onh_crop'][pct]['paths'].append(onh_path)
                processed_data['onh_crop'][pct]['labels'].append(label)
                processed_data['onh_crop'][pct]['vcdr'].append(vcdr)
                # Periphery crop policy
                periphery_cropped = self.cropper.periphery_crop_policy(image, centroid, pct)
                periphery_path = os.path.join(cropped_dir, "periphery_crop", f"{int(pct)}pct", f"image_{idx_out:04d}.jpg")

                cv2.imwrite(periphery_path, cv2.cvtColor(periphery_cropped, cv2.COLOR_RGB2BGR))
                processed_data['periphery_crop'][pct]['paths'].append(periphery_path)
                processed_data['periphery_crop'][pct]['labels'].append(label)
                processed_data['periphery_crop'][pct]['vcdr'].append(vcdr)

        print(f"Processing complete. Processed {len(processed_data['full']['paths'])} images.")
        print(f"Results saved to: {cropped_dir}")
        processed_data['index_map'] = kept_to_original

        return processed_data

    def train_all_models(self, processed_data: Dict, train_indices: List[int],
                        val_indices: List[int], test_indices: List[int], include_vcdr: bool = False):
        """Train separate models for each cropping configuration and run Grad-CAM per config"""

        results = {}
        models_dir = os.path.join(self.results_dir, "models")
        os.makedirs(models_dir, exist_ok=True)

        def _update_test_results_file(conf_name: str, model_path: str, test_indices: List[int], processed_data: Dict):
            """Update the test results file with results from the just-trained model"""
            test_results_file = os.path.join(models_dir, "..", "all_models_test_results.txt")

            # Create file with header if it doesn't exist
            if not os.path.exists(test_results_file):
                with open(test_results_file, 'w') as f:
                    f.write("TEST RESULTS FOR ALL MODELS\n")
                    f.write("="*50 + "\n")
                    f.write("(Updated incrementally as each model finishes training)\n\n")

            if os.path.exists(model_path):
                # Test the model
                crop_model = "onh_crop" in conf_name.lower() or "periphery_crop" in conf_name.lower()
                test_trainer = GlaucomaTrainer(include_vcdr=include_vcdr, reduce_spatial_bias=crop_model)
                test_trainer.model.load_state_dict(torch.load(model_path, map_location=device))

                # Get test data
                test_paths = [processed_data['full']['paths'][i] for i in test_indices if i < len(processed_data['full']['paths'])]
                test_labels = [processed_data['full']['labels'][i] for i in test_indices if i < len(processed_data['full']['labels'])]

                if test_paths:
                    test_dataset = FundusDataset(test_paths, test_labels, transform=test_trainer.val_transform)
                    test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)
                    test_metrics = test_trainer.evaluate_model(test_loader, show_bar=False)

                    # Append to file
                    with open(test_results_file, 'a') as f:
                        f.write(f"{conf_name}:\n")
                        f.write(f"  AUC: {test_metrics['auc']:.4f}\n")
                        f.write(f"  Accuracy: {test_metrics['accuracy']:.4f}\n")
                        f.write(f"  Precision: {test_metrics['precision']:.4f}\n")
                        f.write(f"  Recall: {test_metrics['recall']:.4f}\n")
                        f.write(f"  F1: {test_metrics['f1']:.4f}\n\n")

                    print(f"Test results for {conf_name} appended to: {test_results_file}")
                    print(f"   AUC: {test_metrics['auc']:.4f} | Acc: {test_metrics['accuracy']:.4f}")
                else:
                    with open(test_results_file, 'a') as f:
                        f.write(f"{conf_name}: NO TEST DATA\n\n")
            else:
                with open(test_results_file, 'a') as f:
                    f.write(f"{conf_name}: MODEL NOT FOUND\n\n")

        # Fixed helper to run gradcam export with proper model loading
        def _export_gradcam(trainer: GlaucomaTrainer, conf_name: str, val_paths: List[str], val_labels: List[int]):
            # Load the trained model weights
            model_path = os.path.join(models_dir, f'best_{conf_name}_model.pth')
            if not os.path.exists(model_path):
                print(f"⚠️ Model weights not found: {model_path}")
                return

            # Create a new model instance and load the trained weights
            crop_model = "onh_crop" in conf_name or "periphery_crop" in conf_name
            gradcam_model = GlaucomaClassificationModel(include_vcdr=include_vcdr, reduce_spatial_bias=crop_model).to(device)
            try:
                gradcam_model.load_state_dict(torch.load(model_path, map_location=device))
                gradcam_model.eval()
                print(f"✅ Loaded trained weights for Grad-CAM: {model_path}")
            except Exception as e:
                print(f"⚠️ Failed to load model weights: {e}")
                return

            xai_dir = os.path.join(self.results_dir, "xai", conf_name)

            # Create a temporary trainer just for Grad-CAM with the loaded model
            temp_trainer = GlaucomaTrainer(include_vcdr=include_vcdr, save_dir=models_dir)
            temp_trainer.model = gradcam_model  # Use the loaded model
            temp_trainer.generate_and_save_gradcams(val_paths, xai_dir, max_images=50, labels=val_labels)
            print(f"✅ Grad-CAM saved to: {xai_dir}")

        # ---- 1) Train on full images ----
        print("Training model on full images...")
        crop_model = False  # Full images don't need spatial bias reduction
        trainer = GlaucomaTrainer(include_vcdr=include_vcdr, save_dir=models_dir, reduce_spatial_bias=crop_model)

        train_paths = [processed_data['full']['paths'][i] for i in train_indices
                      if i < len(processed_data['full']['paths'])]
        train_labels = [processed_data['full']['labels'][i] for i in train_indices
                      if i < len(processed_data['full']['labels'])]
        val_paths = [processed_data['full']['paths'][i] for i in val_indices
                    if i < len(processed_data['full']['paths'])]
        val_labels = [processed_data['full']['labels'][i] for i in val_indices
                    if i < len(processed_data['full']['labels'])]

        print(f"[FULL] train paths: {len(train_paths)}, val paths: {len(val_paths)}", flush=True)

        train_vcdr = [processed_data['full']['vcdr'][i] for i in train_indices
                    if i < len(processed_data['full']['vcdr'])] if include_vcdr else None
        val_vcdr = [processed_data['full']['vcdr'][i] for i in val_indices
                  if i < len(processed_data['full']['vcdr'])] if include_vcdr else None

        if train_paths and val_paths:
            train_dataset = FundusDataset(train_paths, train_labels, train_vcdr, trainer.train_transform)
            val_dataset = FundusDataset(val_paths, val_labels, val_vcdr, trainer.val_transform)

            trainer.train_model(train_dataset, val_dataset, num_epochs=30, model_name="full")
            results['full'] = trainer.evaluate_model(
                DataLoader(val_dataset, batch_size=16), show_bar=False, desc="full-final-val"
            )

            # Grad-CAM for FULL
            _export_gradcam(trainer, "full", val_paths, val_labels)
            # Update test results file immediately after full model finishes
            full_model_path = os.path.join(models_dir, 'best_full_model.pth')
            _update_test_results_file("Full Image", full_model_path, test_indices, processed_data)




        # ---- 2) Train on cropped images (onh_crop: exclude ONH, periphery_crop: keep ONH) ----
        for crop_type in ['onh_crop', 'periphery_crop']:
            results[crop_type] = {}
            for pct in processed_data[crop_type].keys():
                conf_name = f"{crop_type}_{pct}"
                print(f"Training model on {conf_name}...")
                crop_model = "onh_crop" in conf_name or "periphery_crop" in conf_name
                trainer = GlaucomaTrainer(include_vcdr=include_vcdr, save_dir=models_dir, reduce_spatial_bias=crop_model)

                c_train_paths = [processed_data[crop_type][pct]['paths'][i] for i in train_indices
                                  if i < len(processed_data[crop_type][pct]['paths'])]
                c_val_paths = [processed_data[crop_type][pct]['paths'][i] for i in val_indices
                                if i < len(processed_data[crop_type][pct]['paths'])]
                print(f"[{conf_name}] train: {len(c_train_paths)}, val: {len(c_val_paths)}", flush=True)

                if c_train_paths and c_val_paths:
                    c_train_labels = [processed_data[crop_type][pct]['labels'][i] for i in train_indices
                                      if i < len(processed_data[crop_type][pct]['labels'])]
                    c_val_labels   = [processed_data[crop_type][pct]['labels'][i] for i in val_indices
                                      if i < len(processed_data[crop_type][pct]['labels'])]
                    c_train_vcdr   = ([processed_data[crop_type][pct]['vcdr'][i] for i in train_indices
                                      if i < len(processed_data[crop_type][pct]['vcdr'])] if include_vcdr else None)
                    c_val_vcdr     = ([processed_data[crop_type][pct]['vcdr'][i] for i in val_indices
                                      if i < len(processed_data[crop_type][pct]['vcdr'])] if include_vcdr else None)

                    train_dataset = FundusDataset(c_train_paths, c_train_labels, c_train_vcdr, trainer.train_transform)
                    val_dataset   = FundusDataset(c_val_paths,   c_val_labels,   c_val_vcdr,   trainer.val_transform)

                    trainer.train_model(train_dataset, val_dataset, num_epochs=30, model_name=conf_name)
                    results[crop_type][pct] = trainer.evaluate_model(
                        DataLoader(val_dataset, batch_size=16), show_bar=False, desc=f"{conf_name}-final-val"
                    )

                    # Grad-CAM per config
                    _export_gradcam(trainer, conf_name, c_val_paths, c_val_labels)

                    # NEW: Update test results file immediately after this model finishes
                    model_path = os.path.join(models_dir, f'best_{conf_name}_model.pth')
                    display_name = f"{crop_type.replace('_', ' ').title()} {pct}%"
                    _update_test_results_file(display_name, model_path, test_indices, processed_data)


        # Save results to JSON
        results_file = os.path.join(self.results_dir, "training_results.json")
        with open(results_file, 'w') as f:
            import json
            json.dump(results, f, indent=4)
        print(f"Training results saved to: {results_file}")

        self.results = results
        return results


# =============================================================================
# EXPERIMENT RUNNER WITH MULTI-DATASET SUPPORT
# =============================================================================

# ========= General Chaksu diagnostics (helpers) =========
def _norm_stem_generic(name: str) -> str:
    base = os.path.splitext(os.path.basename(str(name)))[0].lower()
    return re.sub(r'[^a-z0-9]+', '', base)

def _first_number(base_name: str) -> str:
    m = re.search(r'\d+', os.path.splitext(base_name)[0])
    return m.group(0) if m else ""

def _file_md5_quick(path: str, chunk_size: int = 1 << 20) -> str:
    h = hashlib.md5()
    with open(path, "rb") as f:
        while True:
            b = f.read(chunk_size)
            if not b: break
            h.update(b)
    return h.hexdigest()

def diagnose_chaksu(dataset_root: str, results_dir: str, chaksu_labels_map: dict | None):
    """
    General diagnostics for Chaksu labeling:
      - scans Chaksu_merged/images + masks
      - scans Chaksu_split (if present) → norm/size/md5 maps
      - writes a CSV report of per-file match stage & assigned label
    """
    out_csv = os.path.join(results_dir, "chaksu_debug_report.csv")
    os.makedirs(results_dir, exist_ok=True)

    # locate Chaksu_merged
    merged_dir = None
    for cand in ["Chaksu_merged", "Chaksu", "chaksu"]:
        p = os.path.join(dataset_root, cand)
        if os.path.isdir(p):
            merged_dir = p; break
    if not merged_dir:
        print("[diag] Chaksu_merged not found; aborting.")
        return

    img_dir = None
    for cand in ["images", "Images", "IMG"]:
        p = os.path.join(merged_dir, cand)
        if os.path.isdir(p):
            img_dir = p; break
    mask_dir = None
    for cand in ["masks", "Masks", "GT"]:
        p = os.path.join(merged_dir, cand)
        if os.path.isdir(p):
            mask_dir = p; break

    print(f"[diag] Chaksu_merged images: {img_dir or '(missing)'} | masks: {mask_dir or '(missing)'}")
    ch_imgs = sorted(glob(os.path.join(img_dir, "*.png"))) if img_dir else []
    ch_masks = sorted(glob(os.path.join(mask_dir, "*.tif"))) if mask_dir else []
    print(f"[diag] Found {len(ch_imgs)} Chaksu images, {len(ch_masks)} masks in merged.")

    # quick mask coverage
    mask_bases = {os.path.splitext(os.path.basename(p))[0].lower() for p in ch_masks}
    img_bases  = [os.path.basename(p) for p in ch_imgs]
    no_mask = [b for b in img_bases if os.path.splitext(b)[0].lower() not in mask_bases]
    if no_mask:
        print(f"[diag] WARNING: {len(no_mask)} images missing .tif masks (first 10): {no_mask[:10]}")

    # CSV norm labels (majority per norm key)
    csv_norm_map = {}
    if chaksu_labels_map:
        for k, v in chaksu_labels_map.items():
            csv_norm_map.setdefault(_norm_stem_generic(k), []).append(int(v))
        for k, vals in list(csv_norm_map.items()):
            pos = sum(vals); neg = len(vals) - pos
            csv_norm_map[k] = 1 if pos >= neg else 0
    print(f"[diag] CSV label norms: {len(csv_norm_map)}")

    # Chaksu_split scan
    split_root = None
    for cand in ["Chaksu_split", "chaksu_split"]:
        p = os.path.join(dataset_root, cand)
        if os.path.isdir(p):
            split_root = p; break

    split_norm_map, split_size_map, split_md5_map = {}, {}, {}
    total_scanned = 0
    if split_root:
        for label_dir, y in [("glaucoma", 1), ("control", 0)]:
            for sub in ["images", "Images"]:
                d = os.path.join(split_root, label_dir, sub)
                if not os.path.isdir(d): continue
                files = glob(os.path.join(d, "*"))
                for fp in files:
                    base = os.path.basename(fp)
                    ns = _norm_stem_generic(base)
                    split_norm_map.setdefault(ns, []).append(y)
                    try:
                        size = os.path.getsize(fp)
                        split_size_map.setdefault(size, []).append(y)
                    except Exception:
                        pass
                    total_scanned += 1
        split_norm_map = {k: (1 if sum(v) >= (len(v)-sum(v)) else 0) for k, v in split_norm_map.items()}
        split_size_map = {k: (1 if sum(v) >= (len(v)-sum(v)) else 0) for k, v in split_size_map.items()}

        # md5 sample (cap to 200)
        md5_count = 0
        for label_dir, y in [("glaucoma", 1), ("control", 0)]:
            for sub in ["images", "Images"]:
                d = os.path.join(split_root, label_dir, sub)
                if not os.path.isdir(d): continue
                for fp in glob(os.path.join(d, "*")):
                    if md5_count >= 200: break
                    try:
                        m = _file_md5_quick(fp)
                        split_md5_map.setdefault(m, []).append(y)
                        md5_count += 1
                    except Exception:
                        pass
        split_md5_map = {k: (1 if sum(v) >= (len(v)-sum(v)) else 0) for k, v in split_md5_map.items()}
        print(f"[diag] Chaksu_split present. norm={len(split_norm_map)}, size={len(split_size_map)}, md5={len(split_md5_map)}, total_scanned={total_scanned}")
    else:
        print("[diag] Chaksu_split not found.")

    # per-image evaluation
    import pandas as _pd
    rows = []
    stage_counter = Counter()
    misses = []

    for ip in ch_imgs:
        base = os.path.basename(ip)
        ns   = _norm_stem_generic(base)
        ns_trim    = re.sub(r'(?:[-_]?scan)?\d+$', '', ns)
        ns_vendor  = re.sub(r'^(bosch|forus|remidio)[-_]+', '', ns_trim or ns)
        ns_eye     = re.sub(r'[_-]?(od|os|ou)$', '', ns_vendor or ns_trim or ns)
        ns_first   = _first_number(base)

        label, stage = None, None

        # CSV path first
        for keyname, keyval in [("csv:norm", ns), ("csv:trim", ns_trim), ("csv:firstnum", ns_first),
                                ("csv:vendor", ns_vendor), ("csv:eye", ns_eye)]:
            if keyval and keyval in csv_norm_map:
                label = csv_norm_map[keyval]; stage = keyname; break

        # Split path
        if label is None and split_root:
            for keyname, keyval in [("split:norm", ns), ("split:trim", ns_trim),
                                    ("split:firstnum", ns_first), ("split:vendor", ns_vendor), ("split:eye", ns_eye)]:
                if keyval and keyval in split_norm_map:
                    label = split_norm_map[keyval]; stage = keyname; break

            if label is None:
                try:
                    sz = os.path.getsize(ip)
                    if sz in split_size_map:
                        label = split_size_map[sz]; stage = "split:size"
                except Exception:
                    pass

            if label is None and len(split_md5_map) > 0:
                try:
                    md5 = _file_md5_quick(ip)
                    if md5 in split_md5_map:
                        label = split_md5_map[md5]; stage = "split:md5"
                except Exception:
                    pass

        if label is None:
            stage = "MISS"
            misses.append(base)

        stage_counter[stage] += 1
        try:
            _sz = os.path.getsize(ip)
        except Exception:
            _sz = -1
        try:
            _md5 = _file_md5_quick(ip)
        except Exception:
            _md5 = "md5_error"

        rows.append({
            "filename": base,
            "norm": ns,
            "trim": ns_trim,
            "firstnum": ns_first,
            "vendor_drop": ns_vendor,
            "eye_drop": ns_eye,
            "size_bytes": _sz,
            "md5": _md5,
            "match_stage": stage,
            "assigned_label": ("" if label is None else int(label))
        })

    _pd.DataFrame(rows).to_csv(out_csv, index=False)
    print(f"[diag] Wrote per-file report: {out_csv}")
    print("[diag] Match summary:")
    for k, v in stage_counter.most_common():
        print(f"   {k:>10}: {v}")
    if misses:
        miss_txt = os.path.join(results_dir, "chaksu_misses_only.txt")
        with open(miss_txt, "w") as f:
            for ip in ch_imgs:
                base = os.path.basename(ip)
                if base not in misses:
                    continue
                try:
                    _sz = os.path.getsize(ip)
                except Exception:
                    _sz = -1
                try:
                    _md5 = _file_md5_quick(ip)
                except Exception:
                    _md5 = "md5_error"
                f.write(f"{base}\tsize={_sz}\tmd5={_md5}\n")
        print(f"[diag] {len(misses)} unlabeled files. First 10: {misses[:10]}")
        print(f"[diag] Miss list saved to: {miss_txt}")
# ======= end diagnostics block =======


class ExperimentRunner:
    """Run the complete multi-dataset pipeline (load → split → crop → train → analyze → saliency)."""

    def __init__(self, dataset_root: str, results_dir: str = "/content/drive/MyDrive/Fundus_results"):
        self.dataset_root = dataset_root
        self.results_dir = results_dir
        self.pipeline = None
        self.dataset_loader = MultiDatasetLoader(dataset_root, results_dir)

        os.makedirs(self.results_dir, exist_ok=True)
        print(f"Results will be saved to: {self.results_dir}")

    # -----------------------------
    # Helpers
    # -----------------------------
    def save_dataset_statistics(self, samples: List[Dict]):
        stats = {}
        for s in samples:
            ds = s['dataset']
            stats.setdefault(ds, {'glaucoma': 0, 'normal': 0, 'total': 0})
            if s['label'] == 1:
                stats[ds]['glaucoma'] += 1
            else:
                stats[ds]['normal'] += 1
            stats[ds]['total'] += 1
        path = os.path.join(self.results_dir, 'dataset_statistics.json')
        with open(path, 'w') as f:
            json.dump(stats, f, indent=4)
        print(f"Dataset statistics saved to: {path}")

    def analyze_results(self, results: Dict):
        if not results:
            print("No results to analyze.")
            return

        rows = []
        if 'full' in results:
            rows.append({'Configuration': 'Full Image', 'Crop Type': 'none', 'Crop Percentage': 0,
                         'AUC': results['full']['auc']})
        for crop_type in ['onh_crop', 'periphery_crop']:
            if crop_type in results:
                for pct, metrics in results[crop_type].items():
                    rows.append({'Configuration': f'{crop_type} {pct}%', 'Crop Type': crop_type,
                                 'Crop Percentage': float(pct), 'AUC': metrics['auc']})

        df = pd.DataFrame(rows).sort_values(['Crop Type', 'Crop Percentage'])
        csv_path = os.path.join(self.results_dir, 'experimental_results.csv')
        df.to_csv(csv_path, index=False)
        print(f"Results saved to: {csv_path}")

        # Plot
        try:
            import matplotlib.pyplot as plt
            plt.figure(figsize=(10, 6))
            if 'full' in results:
                plt.axhline(results['full']['auc'], linestyle='--',
                            label=f"Full (AUC={results['full']['auc']:.3f})")
            for crop_type, marker in [('onh_crop', 'o-'), ('periphery_crop', 's-')]:
                sub = df[df['Crop Type'] == crop_type]
                if not sub.empty:
                    plt.plot(sub['Crop Percentage'], sub['AUC'], marker, label=crop_type)
            plt.xlabel('Crop Percentage (%)'); plt.ylabel('AUC')
            plt.title('Performance vs crop percentage'); plt.grid(True, alpha=0.3); plt.legend()
            fig_path = os.path.join(self.results_dir, 'experimental_results.png')
            plt.savefig(fig_path, dpi=300, bbox_inches='tight'); plt.close()
            print(f"Visualization saved to: {fig_path}")
        except Exception as e:
            print(f"(Plotting skipped) Reason: {e}")

        self.print_statistical_analysis(results)

    @staticmethod
    def print_statistical_analysis(results: Dict):
        print("\nSTATISTICAL ANALYSIS")
        print("-" * 40)
        if 'full' in results:
            print(f"Full Image AUC: {results['full']['auc']:.4f}")
        if 'onh_crop' in results and results['onh_crop']:
            vals = [m['auc'] for m in results['onh_crop'].values()]
            best_k, best_m = max(results['onh_crop'].items(), key=lambda kv: kv[1]['auc'])
            print(f"ONH Crop  mean AUC: {np.mean(vals):.4f} ± {np.std(vals):.4f} | best {best_k}%: {best_m['auc']:.4f}")
        if 'periphery_crop' in results and results['periphery_crop']:
            vals = [m['auc'] for m in results['periphery_crop'].values()]
            best_k, best_m = max(results['periphery_crop'].items(), key=lambda kv: kv[1]['auc'])
            print(f"Periphery mean AUC: {np.mean(vals):.4f} ± {np.std(vals):.4f} | best {best_k}%: {best_m['auc']:.4f}")

    @staticmethod
    def find_best_model(results: Dict) -> Tuple[Optional[str], float]:
        best_auc, best_cfg = -1.0, None
        if 'full' in results:
            best_auc = results['full']['auc']; best_cfg = 'full'
        for ctype in ['onh_crop', 'periphery_crop']:
            if ctype in results:
                for pct, metrics in results[ctype].items():
                    if metrics['auc'] > best_auc:
                        best_auc, best_cfg = metrics['auc'], f'{ctype}_{pct}'
        return best_cfg, best_auc

    # -----------------------------
    # Main entry point
    # -----------------------------
    def run_complete_experiment(
        self,
        train_ratio: float = 0.8,
        segmentation_model_path: Optional[str] = None,
        use_pretrained_masks: bool = True,
        chaksu_labels: Optional[Dict[str, int]] = None,
        crop_percentages: List[float] = [10, 20, 30, 40, 50, 60],
    ):
        """
        Load datasets (with optional Chaksu labels), split, crop, train, analyze, and generate saliency.
        Returns (results_dict, average_saliency_or_None).
        """
        print("="*80)
        print("STARTING COMPLETE OPTIC DISC SEGMENTATION AND GLAUCOMA DETECTION PIPELINE")
        print("="*80)
        print(f"Results directory: {self.results_dir}")



        # Load datasets (classification-style list with masks)
        all_samples = self.dataset_loader.load_all_datasets(include_chaksu_labels=chaksu_labels)

        # NEW: run general Chaksu diagnostics right after loading datasets
        try:
            diagnose_chaksu(self.dataset_root, self.results_dir, chaksu_labels)
        except Exception as e:
            print(f"[diag] Skipped Chaksu diagnostics: {e}")

        if not all_samples:
            print("❌ No samples loaded. Please check dataset paths.")
            return None, None

        # Save stats
        self.save_dataset_statistics(all_samples)

        # Init pipeline
        self.pipeline = GlaucomaPipeline(segmentation_model_path, use_pretrained_masks, self.results_dir)

        # Split indices
        y = np.array([s['label'] for s in all_samples], dtype=np.int64)
        sss = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
        train_idx, temp_idx = next(sss.split(np.zeros(len(y)), y))

        sss2 = StratifiedShuffleSplit(n_splits=1, test_size=0.5, random_state=42)
        val_idx, test_idx = next(sss2.split(np.zeros(len(temp_idx)), y[temp_idx]))
        val_idx = temp_idx[val_idx]
        test_idx = temp_idx[test_idx]

        train_indices, val_indices, test_indices = train_idx.tolist(), val_idx.tolist(), test_idx.tolist()

        train_samples = [all_samples[i] for i in train_indices]
        val_samples   = [all_samples[i] for i in val_indices]

        split_info = {
            'total_samples': len(all_samples),
            'train_samples': len(train_indices),
            'val_samples': len(val_indices),
            'test_samples': len(test_indices),
            'train_indices': train_indices,
            'val_indices': val_indices,
            'test_indices': test_indices,
        }
        split_file = os.path.join(self.results_dir, 'data_split.json')
        with open(split_file, 'w') as f:
            json.dump(split_info, f, indent=4)
        print(f"\nDataset split: {len(train_indices)} training, {len(val_indices)} validation samples")
        print(f"Split information saved to: {split_file}")

        # Stage 1 & 2: Cropping
        print("\n" + "="*50)
        print("STAGE 1 & 2: SKIPPED (using existing cropped images)")
        print("="*50)

        import glob

        cropped_dir = os.path.join(self.results_dir, "cropped_images")
        full_dir    = os.path.join(cropped_dir, "full")
        onh_dir     = os.path.join(cropped_dir, "onh_crop")
        peri_dir    = os.path.join(cropped_dir, "periphery_crop")

        if not (os.path.isdir(full_dir) and os.path.isdir(onh_dir) and os.path.isdir(peri_dir)):
            raise RuntimeError(
                f"Existing cropped data not found at {cropped_dir}. "
                "If you changed RESULTS_DIR, set it back to where crops were saved."
            )

        # discover crop percentages from folder names (e.g., '10pct', '20pct', ...)
        def _discover_pcts(base_dir):
            pcts = []
            try:
                for sub in os.listdir(base_dir):
                    if sub.endswith("pct"):
                        try:
                            pcts.append(float(sub.replace("pct","")))
                        except Exception:
                            pass
            except OSError as e:
                print(f"I/O Error accessing {base_dir}: {e}")
                print("Retrying after remounting drive...")
                from google.colab import drive
                drive.flush_and_unmount()
                drive.mount('/content/drive', force_remount=True)
                # Try again after remount
                try:
                    for sub in os.listdir(base_dir):
                        if sub.endswith("pct"):
                            try:
                                pcts.append(float(sub.replace("pct","")))
                            except Exception:
                                pass
                except OSError:
                    print(f"Still cannot access {base_dir} after remount")
                    raise
            return sorted(pcts)

        # If user passed specific crop_percentages, keep their order; otherwise discover:
        disk_pcts = _discover_pcts(onh_dir)
        if not crop_percentages:
            crop_percentages = disk_pcts
        else:
            # keep only those that actually exist on disk to avoid KeyErrors later
            crop_percentages = [p for p in crop_percentages if str(int(p)) + "pct" in os.listdir(onh_dir)]

        # Rebuild processed_data **without re-cropping**, pairing labels from all_samples
        def _collect(sep_dir):
            # use the index used during saving: image_0000.jpg, image_0001.jpg, ...
            paths = sorted(glob.glob(os.path.join(sep_dir, "image_*.jpg")))
            return paths

        full_paths = _collect(full_dir)
        n_full = len(full_paths)

        if n_full == 0:
            raise RuntimeError(f"No files found in {full_dir}. Cannot resume Stage 3 without crops.")

        # Sanity: we assume the previous run kept ordering of `all_samples` when saving.
        # If counts differ, we continue but warn; training will still work because we use labels from all_samples[:n_full].
        if n_full != len(all_samples):
            print(f"⚠️ Warning: full images count ({n_full}) != loaded samples ({len(all_samples)}). "
                  f"Will align by index up to {min(n_full, len(all_samples))}.")

        # Build processed_data in the same structure expected by training
        processed_data = {
            'full': {'paths': [], 'labels': [], 'vcdr': []},
            'onh_crop': {pct: {'paths': [], 'labels': [], 'vcdr': []} for pct in crop_percentages},
            'periphery_crop': {pct: {'paths': [], 'labels': [], 'vcdr': []} for pct in crop_percentages},
            # We don’t have index_map now; we will assume identity mapping (see below).
            'index_map': list(range(n_full))
        }

        # Fill FULL
        limit = min(n_full, len(all_samples))
        for i in range(limit):
            processed_data['full']['paths'].append(full_paths[i])
            processed_data['full']['labels'].append(int(all_samples[i]['label']))
            processed_data['full']['vcdr'].append(float('nan'))

        # Fill ONH and PERIPHERY per percentage
        for pct in crop_percentages:
            sub = f"{int(pct)}pct"
            onh_paths  = sorted(glob.glob(os.path.join(onh_dir, sub, "image_*.jpg")))
            peri_paths = sorted(glob.glob(os.path.join(peri_dir, sub, "image_*.jpg")))
            if len(onh_paths) < limit or len(peri_paths) < limit:
                print(f"⚠️ Warning: {sub} has fewer files than FULL. Using the common minimum.")
            lim_p = min(limit, len(onh_paths), len(peri_paths))
            for i in range(lim_p):
                y = int(all_samples[i]['label'])
                processed_data['onh_crop'][pct]['paths'].append(onh_paths[i])
                processed_data['onh_crop'][pct]['labels'].append(y)
                processed_data['onh_crop'][pct]['vcdr'].append(float('nan'))
                processed_data['periphery_crop'][pct]['paths'].append(peri_paths[i])
                processed_data['periphery_crop'][pct]['labels'].append(y)
                processed_data['periphery_crop'][pct]['vcdr'].append(float('nan'))

        # Because we rebuilt from disk in the same order as the original loop,
        # we can assume 1:1 mapping between original indices and processed indices.
        proc_train_indices = train_indices
        proc_val_indices   = val_indices

        # Stage 3: Training
        print("\n" + "="*50)
        print("STAGE 3: MODEL TRAINING")
        print("="*50)
        results = self.pipeline.train_all_models(
            processed_data, proc_train_indices, proc_val_indices, test_indices, include_vcdr=False
        )

        # === NEW: Evaluate on TEST set ===
        test_paths  = [processed_data['full']['paths'][i] for i in test_indices if i < len(processed_data['full']['paths'])]
        test_labels = [processed_data['full']['labels'][i] for i in test_indices if i < len(processed_data['full']['labels'])]

        if test_paths:
            eval_trainer = GlaucomaTrainer(include_vcdr=False)  # get torchvision val transforms
            # Load the best model for test evaluation
            test_trainer = GlaucomaTrainer(include_vcdr=False, save_dir=os.path.join(self.results_dir, "models"), reduce_spatial_bias=False)
            best_model_path = os.path.join(self.results_dir, "models", "best_full_model.pth")
            if os.path.exists(best_model_path):
                test_trainer.model.load_state_dict(torch.load(best_model_path, map_location=device))

            test_dataset = FundusDataset(test_paths, test_labels, transform=test_trainer.val_transform)
            test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)
            test_metrics = test_trainer.evaluate_model(test_loader, show_bar=False, desc="test")

            print("\n==============================")
            print(" TEST RESULTS")
            print("==============================")
            print(f"Test AUC      : {test_metrics['auc']:.4f}")
            print(f"Test Accuracy : {test_metrics['accuracy']:.4f}")
            print(f"Test Precision: {test_metrics['precision']:.4f}")
            print(f"Test Recall   : {test_metrics['recall']:.4f}")
            print(f"Test F1       : {test_metrics['f1']:.4f}")

            # store in results dict so it’s written to training_results.json
            results['test'] = test_metrics

        # Stage 4: Analysis
        print("\n" + "="*50)
        print("STAGE 4: RESULTS ANALYSIS")
        print("="*50)
        self.analyze_results(results)


        print("\n" + "="*80)
        print("EXPERIMENT COMPLETED SUCCESSFULLY!")
        print(f"All results saved to: {self.results_dir}")
        print("="*80)

        return results, None

# =============================================================================
# MAIN EXECUTION
# =============================================================================

if __name__ == "__main__":
    DATASET_ROOT = "/content/drive/MyDrive/Fundus Dataset"
    RESULTS_DIR  = "/content/drive/MyDrive/Fundus_results"
    chaksu_labels = None  # or a dict if you have them

    runner = ExperimentRunner(dataset_root=DATASET_ROOT, results_dir=RESULTS_DIR)
    results, saliency_map = runner.run_complete_experiment(
        train_ratio=0.8,
        segmentation_model_path=None,
        use_pretrained_masks=True,
        chaksu_labels=chaksu_labels,
    )
```
